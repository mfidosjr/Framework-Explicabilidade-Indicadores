**Artigo:**  
*Desigualdades Territoriais na Qualidade dos Serviços de Telecomunicações: Uma Análise Interpretável com Clusterização Espacial*  

## Objetivo
O propósito central da preparação foi construir uma **base integrada, ampla e multidimensional** de dados socioeconômicos dos municípios brasileiros.  
Essa base consolida variáveis econômicas, demográficas, geográficas, administrativas e de desenvolvimento humano, permitindo análises exploratórias ricas e a aplicação de técnicas de clusterização.  

A ideia é oferecer um leque variado de variáveis, de modo que, em etapas posteriores, possamos selecionar aquelas que mais contribuem para **diferenciar perfis municipais** e formar **clusters homogêneos com maior potencial explicativo**.

## Etapas da Preparação

1. **Coleta e integração de fontes oficiais**  
   - PIB dos Municípios (2021) – indicadores econômicos setoriais e agregados.  
   - População, Área e Densidade (2021) – estrutura territorial e demográfica.  
   - Urbanização (2022) – proporção da população urbana e rural.  
   - Atlas Brasil / PNUD – IDHM (2010).  
   - Cadastro georreferenciado – latitude/longitude, mesorregiões, microrregiões, hierarquias urbanas.  

   ➝ Todas as fontes foram harmonizadas pelo **código IBGE do município**, garantindo consistência.  

2. **Padronização das variáveis**  
   - Normalização de nomes e tipos de dados.  
   - Conversão de indicadores monetários para valores comparáveis.  
   - Remoção de duplicidades e de linhas agregadas (Brasil e UFs).  

3. **Criação da base integrada (56 variáveis)**  
   - **Identificação geográfica:** códigos e nomes de regiões, estados, municípios.  
   - **Localização administrativa:** latitude, longitude, capital, DDD, SIAFI, fuso horário.  
   - **Economia:** VAB setorial, PIB total, PIB per capita, impostos líquidos.  
   - **Demografia e território:** população, área, densidade demográfica.  
   - **Urbanização:** taxa de urbanização, população urbana e rural.  
   - **Desenvolvimento humano:** IDHM.  
   - **Estrutura produtiva:** principais atividades econômicas municipais.  

---

## Justificativa da Abordagem

- **Amplitude vs. Seletividade**  
  Construímos uma base ampla para permitir que métodos de seleção identifiquem as variáveis mais explicativas.  

- **Clusterização orientada a variáveis latentes**  
  A diversidade de variáveis abre espaço para métodos não supervisionados (K-means, DBSCAN, Hierárquica, SOM), avaliando dimensões mais relevantes na formação de grupos.  

- **Potencial para insights aplicados**  
  - Identificação de **perfis municipais** (agrícolas, industriais, urbanos, vulneráveis).  
  - **Comparação regional** para orientar políticas públicas.  
  - Base para **análises preditivas** e simulações de impacto.  

---

## Exemplos de Uso Posterior
1. **Pré-processamento estatístico** – padronização (z-score, min-max).  
2. **Redução de dimensionalidade** – PCA, t-SNE, UMAP.  
3. **Clusterização** – identificação de perfis socioeconômicos municipais.  
4. **Validação** – cruzamento dos clusters com indicadores externos (IDHM, qualidade de vida, resultados eleitorais).  

---

A preparação resultou em uma **infraestrutura de dados ampla e integrada**, que não só apoia a clusterização, mas também serve como base para **análises exploratórias, relatórios estratégicos e políticas públicas baseadas em evidências**.

```python
# 2. Criar dataframes

# Lista de arquivos
arquivos_esperados = [
    "PIB dos Municipios - base de dados 2010-2021.xlsx",
    "municipios_latlong.xlsx",
    "IBGE_codmun-nommun-pop-area-densid-2021.xlsx",
    "IBGE-urbanizacao-2022.xlsx",
    "idhm-atlasbr.xlsx"
]

# Carregar DataFrames
df_pib = pd.read_excel(arquivos_esperados[0], sheet_name="2021")
df_latlong = pd.read_excel(arquivos_esperados[1], sheet_name="municipios_latlong")
df_popdens  = pd.read_excel(arquivos_esperados[2], sheet_name=0)
df_urban = pd.read_excel(arquivos_esperados[3], sheet_name=0)
df_idhm = pd.read_excel(arquivos_esperados[4], sheet_name="Worksheet")

# Mostrar formatos
print("📊 Formato das tabelas:")
print("PIB:", df_pib.shape)
print("LATLONG:", df_latlong.shape)
print("POP_DENS:", df_popdens.shape)
print("URBAN:", df_urban.shape)
print("IDHM:", df_idhm.shape)
```

```python
# 3. Normalização de colunas nos 5 DataFrames principais

# Função de normalização
def normalize_columns(df: pd.DataFrame):
    """
    Retorna (df_norm, mapping_original_para_normalizado).
    - Converte para minúsculas
    - Troca espaços por _
    - Remove acentos e caracteres especiais
    - Mantém apenas letras, números e _
    """
    mapping = {}
    new_cols = []
    for c in df.columns:
        n = str(c).strip().lower().replace("\n", " ")
        n = (
            n.replace("ã","a").replace("á","a").replace("â","a").replace("à","a")
             .replace("é","e").replace("ê","e").replace("è","e")
             .replace("í","i").replace("ì","i")
             .replace("ó","o").replace("ô","o").replace("ò","o")
             .replace("ú","u").replace("ù","u")
             .replace("ç","c")
        )
        n = re.sub(r"\s+", "_", n)         # espaços -> _
        n = re.sub(r"[^\w_]", "", n)       # remove tudo que não é letra, número ou _
        mapping[c] = n
        new_cols.append(n)
    out = df.copy()
    out.columns = new_cols
    return out, mapping

# Dicionário com os DataFrames originais
dfs = {
    "PIB": df_pib,
    "LATLONG": df_latlong,
    "POP_DENS": df_popdens,
    "URBAN": df_urban,
    "IDHM": df_idhm,
}

# Aplicar normalização
dfs_norm = {}
mappings = {}

for name, df in dfs.items():
    df_norm, mapping = normalize_columns(df)
    dfs_norm[name] = df_norm
    mappings[name] = mapping
    
    # Mostrar resumo
    print(f"\n📂 Dataset: {name}")
    print("Número de colunas:", len(df.columns))
    print("Exemplo de mapping (origem -> normalizado):")
    for orig, norm in list(mapping.items())[:8]:  # mostra os 8 primeiros
        print(f"  {orig} --> {norm}")
```

```python
# 4. Perfilamento simples (nome dos campos)

import json

# Criar dicionário apenas com os nomes de colunas
perfis = {
    "PIB": list(df_pib.columns),
    "LATLONG": list(df_latlong.columns),
    "POP_DENS": list(df_popdens.columns),
    "URBAN": list(df_urban.columns),
    "IDHM": list(df_idhm.columns),
}

# Salvar em JSON
with open("perfil_campos.json", "w", encoding="utf-8") as f:
    json.dump(perfis, f, ensure_ascii=False, indent=2)

print("\n✅ Arquivo 'perfil_campos.json' gerado com sucesso!")
```

```python
# 5. Analise de campos para relacionamento dos arquivos

# Carregar o arquivo de perfil de campos
with open("perfil_campos.json", "r", encoding="utf-8") as f:
    perfil = json.load(f)

# Heurísticas de padrões de campos-chave
padroes = {
    "codigo_municipio": re.compile(r"(c[oó]d.*munic|codigo_ibge|cód\.)", re.IGNORECASE),
    "nome_municipio": re.compile(r"(nome.*munic|municipio|brasil e municipio|territorialidades)", re.IGNORECASE),
    "codigo_uf": re.compile(r"(c[oó]d.*uf|sigla_uf)", re.IGNORECASE)
}

# Identificar campos-chave potenciais
resultados = {}
for dataset, campos in perfil.items():
    resultados[dataset] = {}
    for chave, regex in padroes.items():
        matches = [c for c in campos if regex.search(c)]
        if matches:
            resultados[dataset][chave] = matches

# Mostrar de forma amigável
print("🔑 Campos-chave potenciais por dataset:\n")
for ds, chaves in resultados.items():
    print(f"=== {ds} ===")
    for tipo, campos in chaves.items():
        print(f"  {tipo}: {campos}")
    print()
```

**Abordagem a ser seguida:**

1.	Eleger uma tabela-ponte com nome + UF + codigo_ibge. A melhor é LATLONG (tem nome, UF e código).
2.	Normalizar os nomes (sem acentos, caixa baixa, etc.) em URBAN, IDHM e LATLONG.
3.	Dar match determinístico por (nome_normalizado, UF) quando houver UF.
4.	Se não houver UF, casar por nome_normalizado e marcar ambiguidades (nomes repetidos em UFs diferentes) para revisão.
5.	Depois disso, URBAN_mapeada e IDHM_mapeada ganham codigo_ibge e passam a juntar com PIB, POP_DENS e a própria LATLONG pelo código.

```python
# 1.5.1 - Adicionando novos municipios, com valores interpolados a partir das origens

# ----------------------------
# 1) Definir novos municípios e origens, no padrão "Município (UF)"
# ----------------------------
novos_municipios = [
    {"nome_mun": "Mojuí dos Campos (PA)",  "origens": ["Santarém (PA)"]},
    {"nome_mun": "Pinto Bandeira (RS)",    "origens": ["Bento Gonçalves (RS)"]},
    {"nome_mun": "Paraíso das Águas (MS)", "origens": ["Água Clara (MS)", "Camapuã (MS)", "Costa Rica (MS)"]},
    {"nome_mun": "Balneário Rincão (SC)",  "origens": ["Içara (SC)"]},
    {"nome_mun": "Pescaria Brava (SC)",    "origens": ["Laguna (SC)"]},
]

# Coluna de pesos preferida
PESO_COL = "População total 2012"

# ----------------------------
# 2) Função de média ponderada por coluna numérica (ignorando NaN)
# ----------------------------
def media_ponderada_col(serie_vals: pd.Series, pesos: pd.Series):
    # Considera apenas posições onde há valor e (se existirem) pesos
    mask = serie_vals.notna()
    if not mask.any():
        return np.nan
    v = serie_vals[mask].astype(float)
    if pesos is None:
        return v.mean()
    w = pesos[mask].fillna(0)
    if (w > 0).sum() == 0:
        return v.mean()
    return np.average(v, weights=w)

# ----------------------------
# 3) Construir novas linhas
# ----------------------------
linhas_novas = []
nomes_existentes = set(df_idhm["nome_mun"].astype(str))

for mun in novos_municipios:
    # Validação de origens
    origens_df = df_idhm[df_idhm["nome_mun"].isin(mun["origens"])].copy()
    if origens_df.empty:
        print(f"⚠️ Não encontrei origens no IDHM para {mun['nome_mun']}: {mun['origens']}. Pulando.")
        continue

    # Pesos (se a coluna existir); senão, pesos iguais
    if PESO_COL in origens_df.columns:
        pesos = origens_df[PESO_COL]
    else:
        pesos = pd.Series(np.ones(len(origens_df)), index=origens_df.index)

    # Montar a nova linha
    nova_linha = {}
    for col in df_idhm.columns:
        if col == "nome_mun":
            nova_linha[col] = mun["nome_mun"]
        elif np.issubdtype(df_idhm[col].dtype, np.number):
            nova_linha[col] = media_ponderada_col(origens_df[col], pesos)
        else:
            # Campos não numéricos: copiar do primeiro (mantém consistência de metadados se houver)
            nova_linha[col] = origens_df[col].iloc[0]

    # Evitar duplicar se por acaso já existir
    if mun["nome_mun"] in nomes_existentes:
        print(f"ℹ️ '{mun['nome_mun']}' já existe no df_idhm. Não será adicionado novamente.")
    else:
        linhas_novas.append(nova_linha)

# ----------------------------
# 4) Concatenar e validar
# ----------------------------
df_novos = pd.DataFrame(linhas_novas)
df_idhm_expanded = pd.concat([df_idhm, df_novos], ignore_index=True)

print("Antes:", len(df_idhm))
print("Depois:", len(df_idhm_expanded))
print("Diferença:", len(df_idhm_expanded) - len(df_idhm))

# Preview de indicadores-chaves dos novos
cols_preview = ["nome_mun"] + [c for c in df_idhm.columns if ("IDHM" in c or "População total" in c)]
print("\nNovos adicionados (preview):")
print(df_novos[cols_preview].head(len(df_novos)).to_string(index=False))
```

## 5.1 Bloco 0 — Helpers (normalização de nomes + função de merge com prefixo)

```python
# ======================================================
# BLOCO 0 — HELPERS GERAIS
# ======================================================
# O que faz:
#   • Define funções de apoio ao pipeline de integração.
#   • Mantém consistência entre merges, nomes e auditoria.
# Inclui:
#   - nome_norm_strong(): normalização robusta de nomes de municípios.
#   - merge_prefix_on_key(): merge com prefixo e checagem de cobertura.
#   - resumo_colunas(): exibe resumo compacto das colunas de um DataFrame.
# ======================================================

# ------------------------------------------------------
# Normalização forte de nomes de municípios
# ------------------------------------------------------
STOP = {"de", "da", "do", "das", "dos", "d"}

def nome_norm_strong(s: str) -> str:
    """Normaliza o nome do município para uso em chaves textuais."""
    if pd.isna(s):
        return ""
    s = str(s).lower().strip()
    s = unicodedata.normalize("NFKD", s).encode("ascii", "ignore").decode("utf-8")
    s = s.replace("’", "'").replace("´", "'").replace("`", "'")
    s = s.replace("-", " ").replace("/", " ")
    s = re.sub(r"(\b[dD])\s*'\s*", r"\1 ", s)
    s = re.sub(r"[^a-z0-9 ']+", " ", s)
    s = " ".join(s.split())
    tokens = [t for t in s.split() if t not in STOP]
    return "_".join(tokens)

# ------------------------------------------------------
# Merge com prefixo e validação
# ------------------------------------------------------
def merge_prefix_on_key(
    base: pd.DataFrame,
    df: pd.DataFrame,
    prefix: str,
    key: str = "cod_mun",
    include_cols: Optional[Iterable[str]] = None,
    validate: Optional[str] = "m:1",
) -> pd.DataFrame:
    """Faz merge LEFT de df em base pela chave `key`, prefixando as colunas."""
    if key not in base.columns or key not in df.columns:
        raise KeyError(f"Chave '{key}' ausente em um dos DataFrames.")
    
    # harmonizar tipos
    def _coerce_key(s: pd.Series) -> pd.Series:
        try:
            return pd.to_numeric(s, errors="coerce").astype("Int64")
        except Exception:
            return s.astype(str)
    
    base2, add = base.copy(), df.copy()
    base2[key], add[key] = _coerce_key(base[key]), _coerce_key(df[key])
    
    if base2[key].dtype != add[key].dtype:
        base2[key], add[key] = base[key].astype(str), df[key].astype(str)
    
    # restrição opcional
    if include_cols:
        add = add[[key] + [c for c in include_cols if c in add.columns]]
    
    add = add.drop_duplicates(subset=[key])
    cols_metrics = [c for c in add.columns if c != key]
    cols_metrics_pref = [c if c.startswith(prefix+"__") else prefix+"__"+c for c in cols_metrics]
    
    add_pref = pd.concat(
        [add[[key]].reset_index(drop=True),
         add[cols_metrics].set_axis(cols_metrics_pref, axis=1)], axis=1
    )
    
    out = base2.merge(add_pref, on=key, how="left", validate=validate)
    
    blk_cols = [c for c in out.columns if c.startswith(prefix+"__")]
    cov = out[blk_cols].notna().any(axis=1).sum() if blk_cols else 0
    print(f"[MERGE {prefix}] cols={out.shape[1]} | cobertura={cov}/{len(out)}")
    
    return out

# ------------------------------------------------------
# Resumo compacto de colunas (para inspeção rápida)
# ------------------------------------------------------
def resumo_colunas(df: pd.DataFrame, nome_df: str, max_cols: int = 10):
    """
    Mostra resumo do DataFrame:
      - nome
      - número de colunas
      - primeiros N nomes de colunas (ou todos, se menos de N)
    """
    total = len(df.columns)
    nomes = df.columns.tolist()
    preview = nomes[:max_cols]
    mais = f"... (+{total - max_cols})" if total > max_cols else ""
    print(f"{nome_df}: {total} colunas — ({', '.join(preview)}{mais})")
```

## 5.2 -  Bloco — Análise e registro das colunas do df_pib Criação do df base

```python
# ======================================================
# BLOCO 1 — Análise do df_pib e criação da BASE
# Objetivo:
#   • Inspecionar rapidamente as colunas do df_pib (contagem e nomes).
#   • Criar o DataFrame 'base' a partir do df_pib, mantendo TODAS as colunas.
#   • Garantir a presença de chaves essenciais: cod_mun, nome_mun e uf.
#   • Remover duplicatas por cod_mun e reportar linhas/colunas finais.
# Observação:
#   • 'base' será o DataFrame canônico a ser enriquecido nos próximos merges.
# ======================================================

import pandas as pd

# 1) Inspeção rápida do df_pib
resumo_colunas(df_pib, "df_pib", max_cols=12)

# 2) Sanidade mínima de chaves esperadas
if "cod_mun" not in df_pib.columns:
    raise KeyError("df_pib precisa conter a coluna 'cod_mun'. Verifique a etapa de carregamento/normalização.")

if "nome_mun" not in df_pib.columns:
    # Tente detectar nome original e renomear (ajuste se necessário)
    cand = [c for c in df_pib.columns if "Nome do Município" in c or c.lower().strip() == "nome"]
    if cand:
        df_pib = df_pib.rename(columns={cand[0]: "nome_mun"})
    else:
        raise KeyError("df_pib não possui 'nome_mun' nem coluna equivalente detectável.")

# 3) Garantir coluna 'uf'
if "uf" not in df_pib.columns:
    # Tente usar a coluna "Sigla da Unidade da Federação"
    if "Sigla da Unidade da Federação" in df_pib.columns:
        df_pib = df_pib.rename(columns={"Sigla da Unidade da Federação": "uf"})
    else:
        # Derivar UF pelos 2 primeiros dígitos do cod_mun
        coduf_map = {
            11:"RO",12:"AC",13:"AM",14:"RR",15:"PA",16:"AP",17:"TO",
            21:"MA",22:"PI",23:"CE",24:"RN",25:"PB",26:"PE",27:"AL",28:"SE",29:"BA",
            31:"MG",32:"ES",33:"RJ",35:"SP",
            41:"PR",42:"SC",43:"RS",
            50:"MS",51:"MT",52:"GO",53:"DF"
        }
        df_pib = df_pib.copy()
        df_pib["uf"] = pd.to_numeric(df_pib["cod_mun"], errors="coerce").astype("Int64").astype(str).str[:2]
        df_pib["uf"] = pd.to_numeric(df_pib["uf"], errors="coerce").map(coduf_map)

# 4) Criar BASE com todas as colunas do df_pib
base = df_pib.copy()

# 5) Remover duplicatas por cod_mun (mantém a 1ª ocorrência)
linhas_antes = len(base)
base = base.drop_duplicates(subset=["cod_mun"]).reset_index(drop=True)
linhas_depois = len(base)

# 6) Relatórios de consistência
print("\n[BASE] Consistência inicial")
print(f" - Linhas (antes): {linhas_antes}")
print(f" - Linhas (depois de drop_duplicates por cod_mun): {linhas_depois}")
print(f" - Colunas: {len(base.columns)}")
print(" - Tem 'cod_mun'? ", "cod_mun" in base.columns)
print(" - Tem 'nome_mun'? ", "nome_mun" in base.columns)
print(" - Tem 'uf'? ", "uf" in base.columns)

# 7) (Opcional) Estatísticas rápidas da chave
n_unicos = base["cod_mun"].nunique()
n_nulos  = base["cod_mun"].isna().sum()
print(f"\n[BASE] cod_mun — únicos: {n_unicos} | nulos: {n_nulos}")

# 8) (Opcional) Guardar snapshot dos nomes de colunas para auditoria
colunas_iniciais_base = base.columns.tolist()
# with open("colunas_iniciais_base.json", "w", encoding="utf-8") as f:
#     import json; json.dump(colunas_iniciais_base, f, ensure_ascii=False, indent=2)

print("\n✅ BASE criada a partir do df_pib (todas as colunas mantidas). Pronta para os merges.")
```

```python
# ======================================================
# BLOCO 4 — Merge URBAN (uf + nome normalizado) + PATCH documentado
# Objetivo:
#   • Integrar df_urban_fixed usando chave textual robusta (uf + nome_norm_strong).
#   • Evitar duplicidade (_x/_y) limpando URB__* antes do merge.
#   • Diagnosticar faltantes e aplicar patch pontual conhecido.
# Pré-requisitos:
#   • base criada (Blocos 1–3) com 'uf' e 'nome_mun'
#   • df_urban_fixed contém 'uf' e 'nome_mun'
#   • função nome_norm_strong disponível (Bloco 0)
# ======================================================

# 0) Limpeza preventiva: remover colunas URB__ antigas
base = base.drop(columns=[c for c in base.columns if c.startswith("URB__")], errors="ignore")

# 1) Sanidade + resumo
resumo_colunas(df_urban_fixed, "df_urban_fixed", max_cols=8)
for col in ("uf", "nome_mun"):
    assert col in df_urban_fixed.columns, f"df_urban_fixed precisa ter '{col}'."

cols_antes = len(base.columns)

# 2) Criar chave normalizada nos dois lados
b = base.copy()
b["_key_norm"] = b["uf"].astype(str) + "_" + b["nome_mun"].apply(nome_norm_strong)

u = df_urban_fixed.drop_duplicates(subset=["uf","nome_mun"]).copy()
u["_key_norm"] = u["uf"].astype(str) + "_" + u["nome_mun"].apply(nome_norm_strong)

# 3) Prefixar apenas métricas/atributos do URBAN (preservar chave)
cols_metrics_urb = [c for c in u.columns if c not in ["_key_norm","uf","nome_mun"]]
urb_pref = pd.concat([u[["_key_norm"]], u[cols_metrics_urb].add_prefix("URB__")], axis=1)

# 4) Merge por chave normalizada
base = b.merge(urb_pref, on="_key_norm", how="left").drop(columns="_key_norm")

# 5) Consistência e cobertura
cov_urb = base.filter(like="URB__").notna().any(axis=1).sum()
print(f"[MERGE URBAN] cols={len(base.columns)} | cobertura={cov_urb}/{len(base)}")
print("\n[CONSISTÊNCIA URBAN]")
print(f" - Linhas: {len(base)} (esperado: 5570)")
print(f" - Colunas antes: {cols_antes} | depois: {len(base.columns)} | adicionadas: {len(base.columns) - cols_antes}")

# 6) Diagnóstico de faltantes
faltantes_mask = ~base.filter(like="URB__").notna().any(axis=1)
faltantes = base.loc[faltantes_mask, ["cod_mun","uf","nome_mun"]].copy()
print(f"\n[URBAN] Faltantes: {len(faltantes)}")
if len(faltantes) > 0:
    print(faltantes.head(12).to_string(index=False))

# 7) PATCH pontual documentado (aplicado somente se necessário)
#    Caso conhecido: "Graccho Cardoso" (URBAN) -> "Gracho Cardoso" (BASE), UF = SE
if len(faltantes) > 0:
    # checar se o caso "Gracho Cardoso (SE)" está entre os faltantes
    precisa_patch_gc = ((faltantes["uf"] == "SE") & (faltantes["nome_mun"].str.strip().str.lower() == "gracho cardoso")).any()
    if precisa_patch_gc:
        # aplicar correção no df_urban_fixed
        mask_gc = (
            df_urban_fixed["uf"].eq("SE") &
            df_urban_fixed["nome_mun"].str.replace(r"\s+"," ", regex=True).str.strip().str.lower().eq("graccho cardoso")
        )
        n_corr = mask_gc.sum()
        if n_corr > 0:
            df_urban_fixed.loc[mask_gc, "nome_mun"] = "Gracho Cardoso"
            print(f"\n[PATCH URBAN] Corrigido nome em df_urban_fixed: 'Graccho Cardoso' → 'Gracho Cardoso' (SE). Linhas alteradas: {n_corr}")

            # --- refazer APENAS o merge URBAN ---
            base = base.drop(columns=[c for c in base.columns if c.startswith("URB__")], errors="ignore")

            b = base.copy()
            b["_key_norm"] = b["uf"].astype(str) + "_" + b["nome_mun"].apply(nome_norm_strong)

            u = df_urban_fixed.drop_duplicates(subset=["uf","nome_mun"]).copy()
            u["_key_norm"] = u["uf"].astype(str) + "_" + u["nome_mun"].apply(nome_norm_strong)

            cols_metrics_urb = [c for c in u.columns if c not in ["_key_norm","uf","nome_mun"]]
            urb_pref = pd.concat([u[["_key_norm"]], u[cols_metrics_urb].add_prefix("URB__")], axis=1)

            base = b.merge(urb_pref, on="_key_norm", how="left").drop(columns="_key_norm")

            # cobertura final
            cov_urb = base.filter(like="URB__").notna().any(axis=1).sum()
            print(f"[MERGE URBAN • refeito após PATCH] cobertura={cov_urb}/{len(base)}")

            # confirmar linha corrigida
            print(
                base.loc[base["nome_mun"].str.contains("Gracho Cardoso", case=False, na=False),
                         ["cod_mun","uf","nome_mun","URB__Total","URB__Urbana","URB__Rural"]]
                .to_string(index=False)
            )
        else:
            print("\n[PATCH URBAN] Aviso: não encontrei 'Graccho Cardoso' no df_urban_fixed para corrigir.")
```

## 5.6 - Checking Point dos Merges

```python
# ======================================================
# CHECKPOINT — Evolução dos merges (PIB → POP → LAT → URB)
# Objetivo:
#   • Resumir a evolução da BASE após as integrações feitas até agora.
#   • Mostrar: nº de linhas, nº de colunas por bloco, colunas adicionadas,
#     cobertura por dataset (linhas com pelo menos 1 valor no bloco).
# Pré-requisitos:
#   • 'base' já contém: colunas originais do PIB (sem prefixo) + POP__* + LAT__* + URB__*
#   • Função resumo_colunas (opcional) e merges anteriores executados.
# ======================================================

import pandas as pd

# 1) Definir prefixos dos blocos já integrados
PREFIXOS = {
    "POP": "POP__",
    "LAT": "LAT__",
    "URB": "URB__",
    # "IDHM": "IDHM__"  # (ainda não integrado neste checkpoint)
}

# 2) Inferir colunas do núcleo PIB (sem prefixo)
def colunas_pib_nucleo(df: pd.DataFrame) -> list:
    pref_list = list(PREFIXOS.values())
    return [c for c in df.columns if not any(c.startswith(p) for p in pref_list)]

cols_pib_core = colunas_pib_nucleo(base)

# 3) Funções auxiliares para contagem/cobertura por bloco
def cols_do_bloco(df: pd.DataFrame, prefixo: str) -> list:
    return [c for c in df.columns if c.startswith(prefixo)]

def cobertura_bloco(df: pd.DataFrame, prefixo: str) -> int:
    cols = cols_do_bloco(df, prefixo)
    return df[cols].notna().any(axis=1).sum() if cols else 0

# 4) Montar tabela-resumo
linhas = []
total_linhas = len(base)

# Linha do núcleo PIB (sem prefixos)
linhas.append({
    "Bloco": "PIB (núcleo)",
    "Linhas": total_linhas,
    "Colunas do bloco": len(cols_pib_core),
    "Cobertura (linhas com dados)": total_linhas,   # núcleo está presente em 100%
    "Cobertura (%)": 100.0,
    "Exemplo de colunas": ", ".join(cols_pib_core[:5]) + ("..." if len(cols_pib_core) > 5 else "")
})

# Linhas por cada bloco integrado
for nome, pref in PREFIXOS.items():
    cols_blk = cols_do_bloco(base, pref)
    cov     = cobertura_bloco(base, pref)
    linhas.append({
        "Bloco": nome,
        "Linhas": total_linhas,
        "Colunas do bloco": len(cols_blk),
        "Cobertura (linhas com dados)": cov,
        "Cobertura (%)": round((cov / total_linhas) * 100, 2) if total_linhas else 0.0,
        "Exemplo de colunas": ", ".join(cols_blk[:5]) + ("..." if len(cols_blk) > 5 else "")
    })

resumo = pd.DataFrame(linhas)

# 5) Adicionar coluna de "Colunas acumuladas" para visualizar evolução
#    (PIB + POP + LAT + URB, na ordem apresentada na tabela)
resumo["Colunas acumuladas"] = resumo["Colunas do bloco"].cumsum()
```

## 5.7 - Integrando df idhm_expanded

```python
# ======================================================
# BLOCO 5 — Merge IDHM (df_idhm_final)
# Objetivo:
#   • Integrar df_idhm_final à BASE.
#   • Preferir merge por 'cod_mun'; fallback por (uf + nome_mun) normalizado.
#   • Evitar _x/_y limpando IDHM__* antes do merge e deduplicando a chave.
# Pré-requisitos:
#   • base já contém PIB + POP + LAT + URB
#   • df_idhm_final disponível (colunas podem variar: cod_mun/uf/nome_mun)
#   • função nome_norm_strong disponível (Bloco 0)
# ======================================================

# 0) Limpeza preventiva: remover colunas IDHM__ antigas
base = base.drop(columns=[c for c in base.columns if c.startswith("IDHM__")], errors="ignore")

# 1) Resumo e sanidade mínima
resumo_colunas(df_idhm_final, "df_idhm_final", max_cols=8)
assert "nome_mun" in df_idhm_final.columns, "df_idhm_final precisa ter 'nome_mun'."

cols_antes = len(base.columns)

# ==========================
# Caminho A — por cod_mun
# ==========================
if "cod_mun" in df_idhm_final.columns:
    idhm_add = df_idhm_final.copy()

    # Harmonizar tipo da chave
    base["cod_mun"]     = pd.to_numeric(base["cod_mun"], errors="coerce").astype("Int64")
    idhm_add["cod_mun"] = pd.to_numeric(idhm_add["cod_mun"], errors="coerce").astype("Int64")

    # Deduplicar por cod_mun
    idhm_add = idhm_add.drop_duplicates(subset=["cod_mun"]).copy()

    # Prefixar métricas/atributos (tudo exceto cod_mun)
    cols_metrics_idhm = [c for c in idhm_add.columns if c != "cod_mun"]
    idhm_pref = pd.concat(
        [idhm_add[["cod_mun"]], idhm_add[cols_metrics_idhm].add_prefix("IDHM__")],
        axis=1
    )

    # Merge 1:1 por cod_mun
    base = base.merge(idhm_pref, on="cod_mun", how="left", validate="1:1")
    caminho = "cod_mun"

# =======================================
# Caminho B — por (uf + nome) normalizado
# =======================================
else:
    idhm_add = df_idhm_final.copy()

    # Se não houver 'uf', tentar extrair do sufixo "(UF)" em nome_mun
    if "uf" not in idhm_add.columns or idhm_add["uf"].isna().all():
        idhm_add["uf"] = idhm_add["nome_mun"].str.extract(r"\((\w{2})\)\s*$")[0]
        idhm_add["nome_mun"] = idhm_add["nome_mun"].str.replace(r"\s*\(\w{2}\)\s*$", "", regex=True).str.strip()

    assert "uf" in idhm_add.columns, "Não foi possível obter a coluna 'uf' no df_idhm_final."

    # Criar chaves normalizadas nos dois lados
    b = base.copy()
    b["_key_norm"]      = b["uf"].astype(str) + "_" + b["nome_mun"].apply(nome_norm_strong)
    idhm_add["_key_norm"] = idhm_add["uf"].astype(str) + "_" + idhm_add["nome_mun"].apply(nome_norm_strong)

    # Deduplicar por chave
    idhm_add = idhm_add.drop_duplicates(subset=["_key_norm"]).copy()

    # Prefixar métricas
    cols_metrics_idhm = [c for c in idhm_add.columns if c not in ["_key_norm","uf","nome_mun"]]
    idhm_pref = pd.concat([idhm_add[["_key_norm"]], idhm_add[cols_metrics_idhm].add_prefix("IDHM__")], axis=1)

    # Merge
    base = b.merge(idhm_pref, on="_key_norm", how="left").drop(columns="_key_norm")
    caminho = "(uf+nome_norm)"

# ============== Relatórios
cov_idhm = base.filter(like="IDHM__").notna().any(axis=1).sum()
print(f"[MERGE IDHM] via {caminho} | cols={len(base.columns)} | cobertura={cov_idhm}/{len(base)}")

print("\n[CONSISTÊNCIA IDHM]")
print(f" - Linhas: {len(base)} (esperado: 5570)")
print(f" - Colunas antes: {cols_antes} | depois: {len(base.columns)} | adicionadas: {len(base.columns) - cols_antes}")

# Faltantes (sem nenhum valor IDHM__)
faltantes_mask = ~base.filter(like="IDHM__").notna().any(axis=1)
faltantes = base.loc[faltantes_mask, ["cod_mun","uf","nome_mun"]].copy()
print(f"\n[IDHM] Faltantes: {len(faltantes)}")
if len(faltantes) > 0:
    print(faltantes.head(20).to_string(index=False))
```

### 5.7.1 - Correção de inconsistências

```python
# ===========================================
# PATCH DE NOMES (IDHM -> BASE) + RE-MERGE (CORRIGIDO)
# ===========================================
import pandas as pd
import unicodedata, re

# fallback da normalização
if "nome_norm_strong" not in globals():
    def nome_norm_strong(s: str) -> str:
        STOP = {"de","da","do","das","dos"}
        if pd.isna(s): return ""
        s = str(s).lower().strip()
        s = unicodedata.normalize("NFKD", s).encode("ascii","ignore").decode("utf-8")
        s = s.replace("’","'").replace("´","'").replace("`","'")
        s = s.replace("-", " ").replace("/", " ")
        s = re.sub(r"[^a-z0-9 ']+", " ", s)
        s = " ".join(s.split())
        tokens = [t for t in s.split() if t not in STOP]
        return "_".join(tokens)

# 🔧 Corrija o dicionário: NUNCA troque Tupirama->Tabocão
PATCHES_IDHM_TO_BASE = {
    # históricos/ortografia
    ("PA","Santa Isabel do Pará"): "Santa Izabel do Pará",
    # ("TO","Tupirama"): "Tabocão",  # ❌ REMOVIDO (estava invertido)
    ("TO","Tabocão"): "Fortaleza do Tabocao",    # ✅ CORRETO
    ("CE","Itapaje"): "Itapajé",
    ("RN","Augusto Severo"): "Campo Grande",
    ("RN","Presidente Juscelino"): "Serra Caiada",
    ("PB","Santarém"): "Joca Claudino",
    ("PB","Seridó"): "São Vicente do Seridó",
    ("PE","Iguaraci"): "Iguaracy",
    ("BA","Santa Teresinha"): "Santa Terezinha",
    ("MG","Brazopolis"): "Brazópolis",
    ("MG","Dona Eusebia"): "Dona Euzébia",
    ("MG","São Thomé das Letras"): "São Tomé das Letras",
    ("SP","Embu"): "Embu das Artes",
    ("SP","Florinea"): "Florínea",
    ("SP","São Luís do Paraitinga"): "São Luiz do Paraitinga",
    ("MT","Poxoreo"): "Poxoréu",
    # ajustes finos
    ("CE","Itapagé"): "Itapajé",
    ("MG","Brasópolis"): "Brazópolis",
    ("MG","Dona Eusébia"): "Dona Euzébia",
    ("SP","Florínia"): "Florínea",
    ("MT","Poxoréo"): "Poxoréu",
}

def aplicar_patches(df, patches):
    df = df.copy()
    df["nome_mun"] = df["nome_mun"].astype(str).str.strip()
    mask = df.apply(lambda r: (r["uf"], r["nome_mun"]) in patches, axis=1)
    if mask.any():
        df.loc[mask, "nome_mun"] = df.loc[mask].apply(lambda r: patches[(r["uf"], r["nome_mun"])], axis=1)
    return df

def merge_idhm_enxuto(base, df_idhm):
    base = base.drop(columns=[c for c in base.columns if c.startswith("IDHM__")], errors="ignore")
    b = base.copy()
    b["_key_norm"] = b["uf"].astype(str).str.upper().str.strip() + "_" + b["nome_mun"].apply(nome_norm_strong)

    idhm = df_idhm.copy()
    idhm["uf"] = idhm["uf"].astype(str).str.upper().str.strip()
    idhm["nome_mun"] = idhm["nome_mun"].astype(str).str.strip()
    idhm["_key_norm"] = idhm["uf"] + "_" + idhm["nome_mun"].apply(nome_norm_strong)
    idhm = idhm.drop_duplicates(subset=["_key_norm"])

    cols_metrics = [c for c in idhm.columns if c not in ["_key_norm","uf","nome_mun","cod_mun"]]
    idhm_pref = pd.concat([idhm[["_key_norm"]], idhm[cols_metrics].add_prefix("IDHM__")], axis=1)

    cols_antes = len(base.columns)
    merged = b.merge(idhm_pref, on="_key_norm", how="left").drop(columns="_key_norm")

    cov = merged.filter(like="IDHM__").notna().any(axis=1).sum()
    print(f"[MERGE IDHM] cols={len(merged.columns)} | cobertura={cov}/{len(merged)} | +{len(merged.columns)-cols_antes} colunas")

    falt = merged.loc[~merged.filter(like="IDHM__").notna().any(axis=1), ["cod_mun","uf","nome_mun"]]
    print(f"[IDHM] Faltantes: {len(falt)}")
    if not falt.empty:
        print(falt.head(10).to_string(index=False))
    return merged

# 👉 Aplicar patch e re-merge
df_idhm_patched = aplicar_patches(df_idhm_final, PATCHES_IDHM_TO_BASE)
base = merge_idhm_enxuto(base, df_idhm_patched)

# (Opcional) assert para garantir que Tupirama (TO) casou
ok_tupi = (base["uf"].eq("TO") & base["nome_mun"].eq("Tupirama") & base.filter(like="IDHM__").notna().any(axis=1)).any()
print("✅ Tupirama(TO) casou com IDHM?" , bool(ok_tupi))
```

## 5.8 - Checking Point Final

```python
# ======================================================
# CHECKPOINT — Evolução dos merges (CORE → LAT → POP → PIB → URB → IDHM)
# CORE = apenas chaves mínimas; LAT detecta pref/puro; PIB = restante sem prefixo
# ======================================================
import pandas as pd
from textwrap import shorten

# 1) Definições
CORE_KEYS = ["cod_mun", "uf", "nome_mun"]
LAT_RAW   = {"lat","lon","capital","ddd","siafi_id","fuso_horario"}  # possíveis cols LAT sem prefixo
PREFIXOS  = {"POP":"POP__", "URB":"URB__", "IDHM":"IDHM__", "PIB":"PIB__", "LAT":"LAT__"}  # prefixos já padronizados

def has_col(df, c): return c in df.columns
def cols_start(df, p): return [c for c in df.columns if c.startswith(p)]

# 2) Núcleo CORE (apenas chaves)
core_cols = [c for c in CORE_KEYS if c in base.columns]
# 2.1) LAT — pode vir com prefixo (LAT__) ou cru (lat, lon, …)
lat_cols_pref = cols_start(base, PREFIXOS["LAT"])
lat_cols_raw  = [c for c in LAT_RAW if c in base.columns]
lat_cols = sorted(set(lat_cols_pref + lat_cols_raw))

# 3) Colunas sem prefixo (candidatas a serem do PIB puro)
pref_list = [p for k,p in PREFIXOS.items()]
non_pref = [c for c in base.columns if not any(c.startswith(p) for p in pref_list)]
# 3.1) Remover CORE e LAT cru do conjunto não-prefixado → o que sobrar é PIB puro
pib_puro = [c for c in non_pref if c not in set(core_cols) | set(lat_cols_raw)]

# 4) Blocos com prefixo
pop_cols  = cols_start(base, PREFIXOS["POP"])
urb_cols  = cols_start(base, PREFIXOS["URB"])
idhm_cols = cols_start(base, PREFIXOS["IDHM"])
# (PIB pode ser misto: sem prefixo = pib_puro; com prefixo = PIB__*)
pib_pref  = cols_start(base, PREFIXOS["PIB"])

# 5) Função cobertura
def cobertura(cols):
    if not cols: return 0
    return base[cols].notna().any(axis=1).sum()

# 6) Monta a tabela na ordem desejada
linhas = []
total = len(base)

def add_row(nome, cols):
    if not cols: 
        return
    linhas.append({
        "Bloco": nome,
        "Linhas": total,
        "Cols do bloco": len(cols),
        "Cobertura (linhas)": cobertura(cols),
        "Cobertura (%)": round(cobertura(cols)/total*100, 2) if total else 0.0,
        "Exemplos": ", ".join(cols[:5]) + ("..." if len(cols) > 5 else "")
    })

# CORE (chaves)
add_row("CORE (chaves mínimas)", core_cols)

# LAT (pref +/ou cru)
add_row("LAT", lat_cols)

# POP
add_row("POP", pop_cols)

# PIB (puro sem prefixo + com prefixo, se existir)
pib_cols_total = pib_puro + pib_pref
add_row("PIB", pib_cols_total)

# URB
add_row("URB", urb_cols)

# IDHM
add_row("IDHM", idhm_cols)

resumo = pd.DataFrame(linhas)
resumo["Cols acumuladas"] = resumo["Cols do bloco"].cumsum()

# 7) Estilo
resumo_fmt = resumo.copy()
resumo_fmt["Exemplos"] = resumo_fmt["Exemplos"].apply(lambda s: shorten(s, width=120, placeholder="..."))

fmt = {
    "Linhas": "{:,.0f}".format,
    "Cols do bloco": "{:,.0f}".format,
    "Cobertura (linhas)": "{:,.0f}".format,
    "Cobertura (%)": "{:.2f}".format,
    "Cols acumuladas": "{:,.0f}".format,
}

styled = (
    resumo_fmt.style
    .set_table_styles(
        [
            {"selector":"table","props":[("border-collapse","collapse"),("width","100%")]},
            {"selector":"th, td","props":[("border","1px solid #bbb"),("padding","6px 8px"),("vertical-align","top")]},
            {"selector":"thead th","props":[("background-color","#f2f2f2"),("font-weight","600")]}
        ],
        overwrite=True
    )
    .set_properties(**{"text-align":"left"})
    .hide(axis="index")
    .format(fmt, na_rep="—")
    .apply(lambda s: ["background-color: #fcfcfc" if i % 2 else "" for i in range(len(s))], axis=0)
)

print("📌 CHECKPOINT — Evolução dos merges (CORE → LAT → POP → PIB → URB → IDHM)\n")
print(f"Shape atual da BASE: {base.shape[0]:,} linhas × {base.shape[1]:,} colunas\n")
display(styled)

# (Opcional) faltantes por bloco
for nome, cols in [
    ("LAT", lat_cols),
    ("POP", pop_cols),
    ("PIB", pib_cols_total),
    ("URB", urb_cols),
    ("IDHM", idhm_cols),
]:
    if not cols: 
        continue
    falt = base.loc[~base[cols].notna().any(axis=1), ["cod_mun","uf","nome_mun"]]
    print(f"\n[{nome}] Faltantes: {len(falt)}")
    if not falt.empty:
        print(falt.head(10).to_string(index=False))
```
