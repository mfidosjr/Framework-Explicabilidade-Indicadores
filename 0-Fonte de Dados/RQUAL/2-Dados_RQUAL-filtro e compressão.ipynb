### **Legenda dos Serviços**  
- **SMP**: Serviço Móvel Pessoal  
- **SCM**: Serviço de Comunicação Multimídia  
- **STFC**: Serviço Telefônico Fixo Comutado  
- **SeAC**: Serviço de Acesso Condicionado  

### **Observações**  
- **IND**: Indicadores de qualidade obrigatórios.  
- **INF**: Dados informativos complementares.  
- **IR**: Indicador de Reclamações.

## Code

```python
# Dependencias
import pandas as pd
import os

# pip install pyarrow  # Biblioteca para ler/escrever Parquet
```

```python
# Criação de DF a partir de CSV
df_rqual = pd.read_csv(
    'Tabela_CSV_Indicadores_RQUAL_consolidados20250516.csv', 
    sep=';',                  # Separador de colunas (geralmente ',' ou ';')
    decimal=',',              # Separador decimal (',' para formato brasileiro)
    encoding='utf-8'          # Codificação (pode ser 'latin-1' se houver erros)
)
```

```python
# Número de linhas e colunas
linhas, colunas = df_rqual.shape
print(f"O DataFrame tem {linhas} linhas e {colunas} colunas.")

# Memória total utilizada pelo DataFrame (em MB)
memoria_total = df_rqual.memory_usage(deep=True).sum() / (1024 ** 2)  # Converte para MB
print(f"O DataFrame ocupa {memoria_total:.2f} MB na memória.")

# 3. Tamanho em disco (se salvo em CSV)

# Calcular tamanho do arquivo em MB
tamanho_arquivo_mb = os.path.getsize('Tabela_CSV_Indicadores_RQUAL_consolidados20250516.csv') / (1024 ** 2)  # Bytes para MB
print(f"Tamanho do arquivo CSV: {tamanho_arquivo_mb:.2f} MB")
```

```python
# Imprimir primeiras linhas
df_rqual.head() 
```

```python
# Verifique os nomes exatos das colunas
print(df_rqual.columns.tolist())
```

```python
### Excluir colunas que não serão utilizadas

# Lista das colunas a serem removidas
colunas_remover = [
    'Erro Amostral',
    'Limite Inferior',
    'Limite Superior',
    'nome_mun_norm'
]

# Criar novo DataFrame sem as colunas especificadas
df_rqual1 = df_rqual.drop(columns=colunas_remover, errors='ignore')

# Imprimir primeiras linhas
df_rqual1.head() 
```

```python
# Número de linhas e colunas
linhas, colunas = df_rqual1.shape
print(f"O DataFrame tem {linhas} linhas e {colunas} colunas.")

# Memória total utilizada pelo DataFrame (em MB)
memoria_total = df_rqual1.memory_usage(deep=True).sum() / (1024 ** 2)  # Converte para MB
print(f"O DataFrame ocupa {memoria_total:.2f} MB na memória.")
```

```python
# Conversão para o tipo category
df_rqual1['Serviço'] = df_rqual1['Serviço'].astype('category')
df_rqual1['UF'] = df_rqual1['UF'].astype('category')
df_rqual1['Prestadora'] = df_rqual1['Prestadora'].astype('category')
df_rqual1['Tipo'] = df_rqual1['Tipo'].astype('category')
```

```python
# Número de linhas e colunas
linhas, colunas = df_rqual1.shape
print(f"O DataFrame tem {linhas} linhas e {colunas} colunas.")

# Memória total utilizada pelo DataFrame (em MB)
memoria_total = df_rqual1.memory_usage(deep=True).sum() / (1024 ** 2)  # Converte para MB
print(f"O DataFrame ocupa {memoria_total:.2f} MB na memória.")
```

```python
# Salvar em Parquet com compressão
df_rqual1.to_parquet(
    'dados_rqual_otimizado20250516.parquet',
    compression='gzip',  # Compactação eficiente
    engine='pyarrow'
)
```

```python
# Carregar o DataFrame
df_parquet = pd.read_parquet("dados_rqual_otimizado20250516.parquet")
```

```python
# Número de linhas e colunas
linhas, colunas = df_parquet.shape
print(f"O DataFrame tem {linhas} linhas e {colunas} colunas.")

# Memória total utilizada pelo DataFrame (em MB)
memoria_total = df_parquet.memory_usage(deep=True).sum() / (1024 ** 2)  # Converte para MB
print(f"O DataFrame ocupa {memoria_total:.2f} MB na memória.")

# Primeiras linhas
df_parquet.head()
```
