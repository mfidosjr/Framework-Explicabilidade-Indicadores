#### **Explicações-Chave**  
#### 1. **INF4-DL e INF4-UP (Velocidades de Download/Upload)**  
- **Clusterização**: Agrupe municípios por faixas de velocidade (ex.: >100 Mbps vs. <50 Mbps).  
- **Temporal**: Modele a evolução da infraestrutura (ex.: crescimento de 10% ao ano na velocidade média).  
- **Causal**: Avalie se investimentos em fibra óptica causam aumento direto nas velocidades.  
- **Fairness**: Verifique se áreas rurais têm velocidades sistematicamente menores que urbanas.  

#### 2. **IND9 (Cumprimento de Prazos)**  
- **Explicabilidade (SHAP)**: Descubra se atrasos são causados por fatores como densidade populacional ou número de prestadoras.  
- **Causal (DML)**: Estime o impacto de políticas públicas (ex.: fiscalização intensiva) na redução de atrasos.  

#### 3. **INF1 (Tempo de Reparo)**  
- **Espacial (GNN)**: Identifique regiões com tempos de reparo anormalmente altos (ex.: Norte vs. Sul).  
- **Fairness**: Cheque se municípios de baixa renda sofrem com tempos de reparo mais longos.

```python
# ===============================
# 🧹 2. Garantir que a coluna 'Resultado' esteja em formato numérico
# ===============================
df_parquet['Resultado'] = pd.to_numeric(df_parquet['Resultado'], errors='coerce')

# ===============================
# 📊 3. Função para gerar tabela comparativa
# ===============================
def gerar_tabela_indicadores(df):
    # Filtrar apenas resultados maiores que zero
    df_filtrado = df[df['Resultado'] > 0]

    # Agrupar por Ano e Indicador, contando municípios únicos
    resultado_agrupado = (
        df_filtrado.groupby(['Ano', 'Indicador'])['Município']
        .nunique()
        .reset_index()
    )

    # Pivotar para formar a tabela (linhas: Ano, colunas: Indicadores)
    tabela = (
        resultado_agrupado.pivot(index='Ano', columns='Indicador', values='Município')
        .fillna(0)
        .astype(int)
    )

    # Adicionar total de indicadores com pelo menos um município no ano
    tabela['Total Indicadores com Dados'] = (tabela > 0).sum(axis=1)

    return tabela
```

### 4 - IQST municipal (Indicador Sintético de Qualidade dos Serviços de Telecomunicações

```python
import pandas as pd
import numpy as np
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler

# ===============================
# 1. Garantir que 'Resultado' está como número
# ===============================
df_parquet['Resultado'] = pd.to_numeric(df_parquet['Resultado'], errors='coerce')

# ===============================
# 2. Filtrar apenas dados de 2022
# ===============================
df_2022 = df_parquet[df_parquet['Ano'] == 2022].copy()

# ===============================
# 3. Calcular média mensal por Município e Indicador
# ===============================
df_medias_mensais = (
    df_2022.groupby(['Município', 'Indicador'])['Resultado']
    .mean()
    .reset_index()
)

# ===============================
# 4. Pivotar: cada indicador vira uma coluna
# ===============================
df_pivot = df_medias_mensais.pivot(index='Município', columns='Indicador', values='Resultado')
df_pivot = df_pivot.reset_index()

# ===============================
# 5. Criar os IQSTs
# ===============================

# (a) IQST 1: Média simples dos indicadores
df_pivot['IQST_média_simples'] = df_pivot.iloc[:, 1:9].mean(axis=1, skipna=True)

# (b) IQST 2: Média ponderada conforme cobertura
cobertura = df_pivot.iloc[:, 1:9].notna().sum() / df_pivot.shape[0]
pesos = cobertura / cobertura.sum()
df_pivot['IQST_média_ponderada'] = df_pivot.iloc[:, 1:9].multiply(pesos, axis=1).sum(axis=1)

# (c) IQST 3: Z-score médio
df_zscore = (df_pivot.iloc[:, 1:9] - df_pivot.iloc[:, 1:9].mean()) / df_pivot.iloc[:, 1:9].std()
df_pivot['IQST_zscore'] = df_zscore.mean(axis=1, skipna=True)

# (d) IQST 4: PCA (com 1 componente principal)
df_pca = df_pivot.dropna(subset=df_pivot.columns[1:9]).copy()
scaler = StandardScaler()
X_scaled = scaler.fit_transform(df_pca.iloc[:, 1:9])
pca = PCA(n_components=1)
df_pca['IQST_PCA'] = pca.fit_transform(X_scaled)

# Adicionar resultados de PCA ao df original
df_pivot = df_pivot.merge(df_pca[['Município', 'IQST_PCA']], on='Município', how='left')

# ===============================
# 6. Tabela final comparativa
# ===============================
tabela_iqst = df_pivot[[
    'Município', 'IQST_média_simples', 'IQST_média_ponderada', 'IQST_zscore', 'IQST_PCA'
]]

from IPython.display import display  # para garantir que funcione fora de notebook

# Identificar as colunas numéricas (exceto 'Município')
colunas_numericas = tabela_iqst.select_dtypes(include='number').columns.tolist()

# Aplicar estilização apenas nas colunas numéricas
styled_iqst = (
    tabela_iqst.style
    .set_caption("Indicador Sintético de Qualidade dos Serviços de Telecomunicações (IQST) – 2022")
    .background_gradient(cmap='Blues', subset=colunas_numericas, axis=1)
    .format("{:.2f}", subset=colunas_numericas)
    .highlight_null(subset=colunas_numericas, props='color: gray; background-color: #f0f0f0;')
)

# Exibir estilizado no Jupyter/Colab
# display(styled_iqst)
```

#### Construção do Indicador Sintético de Qualidade dos Serviços de Telecomunicações (IQST)

Metodologia

Para sintetizar os oito indicadores técnicos de qualidade da ANATEL em um escore único por município (IQST), foram testadas quatro abordagens:

- **Média Simples**: média direta dos indicadores disponíveis por município.
- **Média Ponderada**: ajustada pela cobertura nacional de cada indicador (proporção de municípios com dados válidos).
- **Z-score Médio**: normalização dos indicadores seguida da média padronizada.
- **PCA**: projeção dos dados normalizados no primeiro componente principal (Análise de Componentes Principais).

Dada a presença de dados ausentes (notadamente em INF4-DL e INF4-UP), a **média ponderada** apresentou melhor equilíbrio entre cobertura, interpretabilidade e estabilidade estatística. Essa abordagem foi, portanto, adotada como referência para análise e ranking dos municípios.

A média ponderada foi escolhida como índice principal por combinar interpretabilidade, estabilidade e cobertura nacional.

### 5 - Criação do dataset com IQST

```python
# Criação de um novo DataFrame para os indicadores

# Carregar os dados do arquivo parquet
df_parquet = pd.read_parquet('dados_rqual_otimizado20250516.parquet')

# Garantir que 'Resultado' seja numérico
df_parquet['Resultado'] = pd.to_numeric(df_parquet['Resultado'], errors='coerce')

# Filtrar ano 2022
df_2022 = df_parquet[df_parquet['Ano'] == 2022].copy()

# Agrupar média mensal por município, indicador e Código IBGE
df_medias_mensais = (
    df_2022.groupby(['Município', 'Código IBGE', 'Indicador'])['Resultado']
    .mean()
    .reset_index()
)

# Pivotar: um município por linha, um indicador por coluna
df_pivot = df_medias_mensais.pivot(index=['Município', 'Código IBGE'], columns='Indicador', values='Resultado').reset_index()

# Calcular cobertura e pesos
cobertura = df_pivot.iloc[:, 2:10].notna().sum() / df_pivot.shape[0]
pesos = cobertura / cobertura.sum()

# Calcular IQST (média ponderada)
df_pivot['IQST_média_ponderada'] = df_pivot.iloc[:, 2:10].multiply(pesos, axis=1).sum(axis=1)

# Consolidar DataFrame final
colunas_finais = ['Município', 'Código IBGE', 'IND2', 'IND4', 'IND5', 'IND8', 'IND9', 'INF1', 'INF4-DL', 'INF4-UP', 'IQST_média_ponderada']
df_indicadores = df_pivot[colunas_finais]

# Exportar resultados
df_indicadores.to_csv("Indicadores_IQST_2022_20250516.csv", index=False)
df_indicadores.to_parquet("Indicadores_IQST_2022_20250516.parquet", index=False)

# Exibir amostra
print(df_indicadores.head())
```

#### Explicação da Consolidação dos Indicadores e Cálculo do IQST

Estamos utilizando **médias** em dois momentos distintos, com técnicas e propósitos diferentes:

---

1. Consolidação dos Indicadores por Município e Ano

**Objetivo:** obter um único valor anual por município para cada indicador técnico da ANATEL.

**Técnica utilizada:**  
✔️ **Média simples** dos valores mensais registrados no ano.

📌 Exemplo:  
Se um município possui dados mensais para o indicador `IND5` nos meses de março, junho e setembro, o valor final consolidado de `IND5` será a **média aritmética** desses três meses.

---

2. Cálculo do IQST (Indicador Sintético de Qualidade)

**Objetivo:** criar um escore único por município, consolidando os oito indicadores técnicos.

**Técnica utilizada:**  
✔️ **Média ponderada**, com pesos proporcionais à cobertura (número de municípios com dado disponível para cada indicador).

📌 Justificativa:  
Indicadores como `INF4-DL` e `INF4-UP` estão ausentes em muitos municípios. A média ponderada evita que indicadores com baixa cobertura distorçam o escore final.

| Cálculo do IQST sintético           | Média ponderada     | Compensar ausência de indicadores pouco disponíveis |
