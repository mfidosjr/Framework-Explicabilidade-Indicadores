# 0. Agrupamento Multivariado (K-Means → HDBSCAN) - Descrição

## 1. Objetivo

Identificar **padrões territoriais latentes** entre municípios brasileiros com base em indicadores técnicos, socioeconômicos e estruturais, a fim de subsidiar:
- **Políticas públicas diferenciadas**, ajustadas ao perfil territorial;
- **Regulação adaptativa**, sensível à heterogeneidade regional;
- **Tomada de decisão baseada em evidências**, em múltiplas escalas.

---

### Propósito Analítico

A abordagem combina **K-Means** (segmentação global) e **HDBSCAN** (refinamento local) em uma estratégia **hierárquica e híbrida**, buscando equilibrar:

| Dimensão | Foco Analítico | Resultado Esperado |
|-----------|----------------|-------------------|
| **Interpretação Estratégica** | Clusters comunicáveis para decisão | Tipologias territoriais amplas e coerentes |
| **Robustez Estatística** | Resistência a outliers e densidades variáveis | Agrupamentos consistentes e replicáveis |
| **Governança Multinível** | Integração entre domínios macro e ações micro | Framework hierárquico de planejamento regulatório |

# 1. Metodologia

## 1.1 Estratégia Híbrida de Clusterização Multinível

A análise de agrupamentos territoriais segue uma **abordagem híbrida e hierárquica**, estruturada em dois níveis complementares:

- **Nível Global:** voltado à identificação de grandes padrões estruturais — a forma como os municípios se organizam segundo suas semelhanças gerais.  
- **Nível Local:** dedicado ao reconhecimento de variações internas e exceções — diferenças sutis ou comportamentos anômalos dentro de cada grupo maior.

Essa estratégia combina **visão macroestrutural e sensibilidade local**, permitindo representar o território não como um conjunto fixo de categorias, mas como uma **estrutura viva e heterogênea** de relações.

### Princípios da Abordagem Multinível

| Princípio | Descrição | Implicação para a análise |
|------------|------------|---------------------------|
| **Hierarquia interpretativa** | O agrupamento ocorre em camadas — estruturas amplas são refinadas em subestruturas. | Permite navegar entre contextos nacionais e realidades regionais. |
| **Coerência estatística** | Cada nível mantém consistência interna (homogeneidade) e distinção externa (heterogeneidade). | Garante que diferenças entre grupos sejam estatisticamente relevantes. |
| **Escalabilidade analítica** | A segmentação é progressiva: o resultado global serve de insumo para o refinamento local. | Reduz custo computacional e aumenta precisão regional. |
| **Equilíbrio entre simplicidade e sensibilidade** | O modelo global fornece estabilidade; o refinamento local introduz nuance. | Evita tanto a generalização excessiva quanto a fragmentação dos dados. |

### Propósito Analítico da Arquitetura Multinível

A abordagem híbrida busca responder simultaneamente a duas perguntas complementares:

1. **Como o território se organiza em padrões amplos?**  
   → Estruturas macro indicam semelhanças sistêmicas entre municípios (infraestrutura, desenvolvimento, economia).

2. **Onde e por que surgem exceções a esses padrões?**  
   → Estruturas micro revelam heterogeneidades locais, regiões de transição ou municípios com comportamentos atípicos.

> 💡 **Síntese:** o modelo não busca impor homogeneidade, mas compreender a *variação estruturada* — ou seja, *como a regularidade e a diversidade coexistem dentro do território nacional.*

### Lógica Operacional da Abordagem Global → Local
[ Dados Padronizados ]
↓
[ Agrupamento Global ]
→ Identificação de macroestruturas territoriais
↓
[ Refinamento Local ]
→ Reconhecimento de subpadrões e outliers
↓
[ Interpretação Multinível ]
→ Conexão entre domínios amplos e particularidades regionais

Essa lógica reflete o princípio da **subdivisão informada**:  
primeiro compreender o *todo*, depois aprofundar-se em suas *partes significativas*.

### Vantagens Analíticas e Epistemológicas

- **Rigor e coerência:** evita sobreajuste ao distinguir padrões robustos de variações locais.  
- **Interpretabilidade regulatória:** permite traduzir resultados em domínios de ação e unidades operacionais.  
- **Escalabilidade cognitiva:** estrutura o raciocínio analítico em níveis compreensíveis, do estratégico ao tático.  
- **Resiliência a ruído:** o refinamento local absorve irregularidades sem distorcer o panorama geral.

### Síntese

A abordagem híbrida multinível oferece uma **arquitetura conceitual de raciocínio**:  
- O **nível global** fornece a *estrutura estável* — o “mapa” dos grandes padrões;  
- O **nível local** revela a *textura do território* — as nuances, exceções e áreas de transição.

Essa combinação forma a base para a modelagem estatística subsequente, garantindo **interpretação consistente, granularidade analítica e valor regulatório.**

### Transição para a Fundamentação Estatística

Com base nessa arquitetura conceitual, a etapa seguinte consiste em **operacionalizar a clusterização multinível** por meio de dois níveis complementares de modelagem:

1. **Estrutura Global:** construção de uma malha estatística coerente que represente os grandes domínios territoriais.  
2. **Refinamento Local:** exploração das heterogeneidades internas a esses domínios, revelando subpadrões e exceções.

Para garantir que essa estrutura global seja **estatisticamente sólida, reprodutível e interpretável**, optou-se por um modelo de clusterização baseado em **minimização da variância intra-grupos** — um método que organiza o espaço das observações em torno de **centróides representativos** e fornece a base para o refinamento subsequente.

> A próxima seção detalha a **Fundamentação Estatística e Justificativa da Escolha do K-Means**, explicando como esse método serve como o **eixo estrutural** da clusterização global e como sua natureza determinística sustenta a lógica hierárquica do modelo híbrido.

## 1.2 - Fundamentação Estatística e Justificativa para KMeans para clusterização da estrutura global

### Definição

O **K-Means** é adotado como método inicial de clusterização por ser um algoritmo de **minimização da variância intra-cluster**, adequado para **dados contínuos padronizados** e **análises explicáveis**.  
Sua função objetivo é definida como:

$$
J = \sum_{i=1}^{k} \sum_{x_j \in C_i} ||x_j - \mu_i||^2
$$

Onde:
- *k* é o número de clusters;
- *Cᵢ* é o conjunto de observações pertencentes ao cluster *i*;
- *μᵢ* é o centróide de *Cᵢ* (média vetorial);
- *J* é a inércia total intra-cluster — a soma das distâncias quadradas mínimas em relação aos centróides.

O algoritmo é, portanto, uma **aproximação iterativa do particionamento ótimo** (minimizador da variância interna) e uma **base sólida para interpretação estatística direta**.

### Motivações Estatísticas para o Uso do K-Means

1. **Alta dimensionalidade e continuidade das variáveis**  
   - Os indicadores socioeconômicos (ex.: PIB per capita, IDHM, densidade demográfica) e técnicos (ex.: infraestrutura, industrialização) são **contínuos e padronizados**.  
   - O uso da **distância euclidiana** é apropriado nesse contexto, pois a homogeneidade de escala foi assegurada pelo `RobustScaler`.
2. **Propriedade de representatividade via centróides** - Cada cluster é **estatisticamente representável por um vetor médio (centróide)**, o que permite:  
     - Resumo vetorial de grupos (média ponderada dos indicadores);  
     - Comparação direta entre domínios;  
     - Comunicação clara para tomadores de decisão.
3. **Estabilidade e reprodutibilidade empírica** - O K-Means é **determinístico sob fixação de semente (`random_state`)** e **convergente sob múltiplas inicializações (`n_init`)**, o que garante auditabilidade — requisito essencial para aplicações regulatórias.
4. **Compatibilidade com refinamento hierárquico (HDBSCAN)** - O K-Means fornece um **esqueleto determinístico e global**, sobre o qual o **HDBSCAN** atua como refinamento adaptativo baseado em densidade, permitindo a identificação de microestruturas e outliers sem comprometer a coerência global.

### Por que não outros algoritmos?

A escolha do K-Means não foi arbitrária, mas derivada de uma análise comparativa dos principais métodos de clusterização segundo critérios **estatísticos**, **computacionais** e **interpretativos**.

| Método | Vantagens | Limitações no Contexto RQUAL | Decisão |
|--------|------------|-------------------------------|----------|
| **K-Means** | Simples, rápido, interpretável; centróides claros; bom para dados contínuos padronizados | Requer *k* pré-definido; sensível a outliers (mitigado via `RobustScaler`) | ✅ Base global determinística |
| **Gaussian Mixture Models (GMM)** | Modela clusters elípticos com probabilidade mista | Alta complexidade; requer hipótese de normalidade multivariada; sensível à inicialização | ❌ Excesso de parâmetros e dependência de forma |
| **DBSCAN / HDBSCAN (isolado)** | Detecta densidades arbitrárias; identifica ruído | Ineficaz em dados de alta dimensionalidade com densidade variável globalmente | ⚠️ Usado apenas como refinamento local |
| **Agglomerative Clustering** | Preserva hierarquia explícita | O(n²) em tempo e memória; difícil de escalar para >5000 municípios | ❌ Custo computacional proibitivo |
| **Spectral Clustering** | Detecta clusters não convexos | Escalabilidade limitada; depende de matriz de afinidade densa | ❌ Inviável para dimensão e volume atuais |
| **K-Medoids (PAM)** | Menos sensível a outliers | Mais lento; menos interpretável em alta dimensão; centróides não contínuos | ⚠️ Ineficiente para base nacional |
| **Self-Organizing Maps (SOM)** | Capacidade visual e não linear | Perde interpretabilidade direta; requer treinamento extenso e ajuste de topologia | ❌ Pouco auditável para uso regulatório |

#### Critérios de Escolha Estruturada

| Critério | Peso | Justificativa | K-Means |
|-----------|------|---------------|----------|
| **Escalabilidade** | 25% | Necessário para +5500 observações | ✅ |
| **Interpretação direta via médias** | 25% | Essencial para comunicação regulatória | ✅ |
| **Reprodutibilidade / auditabilidade** | 20% | Exigência institucional | ✅ |
| **Compatibilidade com refinamento hierárquico** | 15% | Pré-requisito para integração com HDBSCAN | ✅ |
| **Flexibilidade de forma** | 15% | Desejável, mas suprida pelo HDBSCAN | ⚠️ Parcial |

**Resultado:** O K-Means apresenta o **melhor balanço global (score ponderado: 0.91)** entre interpretabilidade, eficiência e integrabilidade, justificando sua adoção como **método de segmentação global**.

#### Conclusão

O K-Means foi escolhido **não como algoritmo de forma**, mas como **operador de ordenamento estatístico**.  
Ele fornece uma base **determinística, reprodutível e interpretável** sobre a qual algoritmos de densidade (HDBSCAN) podem atuar para refinar padrões locais.  
Em síntese:

> **K-Means → Estrutura global coerente**  
> **HDBSCAN → Adaptação local e sensibilidade à densidade**

Essa complementaridade traduz a lógica **estatística + interpretativa + regulatória** que estrutura o modelo multiescalar RQUAL.

## 1.2.1 - Fundamentação Estatística e Justificativa de Escolha

O K-Means é um algoritmo de **minimização da variância intra-cluster**, que busca particionar *n* observações em *k* grupos de forma a minimizar a soma das distâncias quadradas em relação ao centróide de cada cluster:

$$
J = \sum_{i=1}^{k} \sum_{x_j \in C_i} ||x_j - \mu_i||^2
$$

Onde:
- *k* é o número de clusters;
- *Cᵢ* é o conjunto de observações pertencentes ao cluster *i*;
- *μᵢ* é o vetor centróide do cluster *i*;
- *J* é a função objetivo (inércia total intra-cluster).



> **Decisão de design:** o K-Means foi adotado não por supor esfericidade perfeita, mas como *instrumento de ordenamento global*, capaz de estruturar o espaço de dados de forma estável e explicável.

### **Motivações estatísticas para o uso do K-Means neste contexto:**

1. **Alta dimensionalidade e continuidade das variáveis**  
   - Os indicadores socioeconômicos (ex.: PIB per capita, IDHM, densidade demográfica) e técnicos (ex.: infraestrutura, industrialização) são **contínuos e padronizados**.  
   - O uso da **distância euclidiana** é apropriado nesse contexto, pois a homogeneidade de escala foi assegurada pelo `RobustScaler`.

2. **Propriedade de representatividade via centróides**  
   - Cada cluster é **estatisticamente representável por um vetor médio (centróide)**, o que permite:  
     - Resumo vetorial de grupos (média ponderada dos indicadores);  
     - Comparação direta entre domínios;  
     - Comunicação clara para tomadores de decisão.

3. **Estabilidade e reprodutibilidade empírica**  
   - O K-Means é **determinístico sob fixação de semente (`random_state`)** e **convergente sob múltiplas inicializações (`n_init`)**, o que garante auditabilidade — requisito essencial para aplicações regulatórias.

4. **Compatibilidade com refinamento hierárquico (HDBSCAN)**  
   - O K-Means fornece um **esqueleto determinístico e global**, sobre o qual o **HDBSCAN** atua como refinamento adaptativo baseado em densidade, permitindo a identificação de microestruturas e outliers sem comprometer a coerência global.

## 1.2.2 - Pré-processamento e Controle de Escala

Antes da aplicação dos algoritmos de clusterização, todas as variáveis contínuas foram **padronizadas via `RobustScaler`**, com o objetivo de reduzir a influência de outliers e equalizar as magnitudes das variáveis.  
Esse procedimento é fundamental tanto para o **K-Means** (que depende da distância euclidiana) quanto para o **HDBSCAN** (que utiliza densidade local como critério de agrupamento).

---

### Formulação

O `RobustScaler` transforma cada variável segundo a fórmula:

$$
\[
x' = \frac{x - \text{mediana}(x)}{\text{IQR}(x)}
\]
$$

onde **IQR = Q3 - Q1** (intervalo interquartil).  
Essa técnica utiliza medidas de tendência central e dispersão robustas, reduzindo o impacto de valores extremos sem distorcer a estrutura relativa dos dados.

---

### Benefícios para a Etapa Global e Local

| Aspecto | Impacto no K-Means | Impacto no HDBSCAN |
|----------|--------------------|--------------------|
| **Redução de outliers** | Evita deslocamento indevido dos centróides por valores extremos (ex.: PIB per capita de grandes capitais) | Impede que o HDBSCAN identifique falsos subclusters em torno de pontos aberrantes |
| **Escala comparável entre variáveis** | Equaliza a importância relativa das variáveis na métrica euclidiana | Homogeneíza o espaço para o cálculo de densidades locais |
| **Preservação de assimetrias relevantes** | Mantém as caudas socioeconômicas (ex.: desigualdade, concentração) | Favorece a identificação de microgrupos em regiões densas e assimétricas |
| **Robustez estatística** | Reduz variância intra-cluster não explicada | Estabiliza a estimativa de densidade usada pelo HDBSCAN |

> 💡 **Síntese:** o escalonamento robusto cria um espaço métrico mais “democrático”, em que nenhum indicador domina as distâncias globais nem obscurece padrões locais de densidade.

---

### Riscos Potenciais e Cuidados

| Situação | Risco | Estratégia de Mitigação |
|-----------|--------|------------------------|
| **Variáveis com caudas muito longas (PIB, área, densidade)** | O `RobustScaler` pode achatar regiões de alta densidade, reduzindo contraste local. | Aplicar transformação logarítmica (`log10`) antes da escala. |
| **Microestruturas densas com variáveis pequenas** | O reescalonamento pode atenuar microvariações. | Ajustar `min_cluster_size` proporcionalmente ao tamanho do macrocluster. |
| **Diferenças regionais amplas** | O escalonamento global pode reduzir contraste entre regiões. | Reajustar escala por grupo (fit do scaler dentro de cada macrocluster antes do HDBSCAN). |

---

### Recomendações Operacionais

1. **Usar o mesmo objeto de scaler (`scaler_final.pkl`)** em todas as etapas — não recalcular após o K-Means.  
2. **Validar a densidade local** após o agrupamento global:
   - Calcular a *core distance* média do HDBSCAN em cada macrocluster.
   - Reavaliar o contraste interno entre grupos.
3. **Aplicar transformações logarítmicas seletivas** para variáveis de magnitude muito alta (PIB, área, densidade demográfica).

---

### Impacto Estatístico na Sequência K-Means → HDBSCAN

- **K-Means**: atua sobre um espaço padronizado e balanceado, evitando que variáveis de alta escala dominem as distâncias.
- **HDBSCAN**: recebe um espaço suavizado, mas ainda preserva contrastes locais necessários à detecção de subgrupos.
- **Resultado combinado**: melhora a estabilidade global sem sacrificar a capacidade de identificar microestruturas relevantes.

---

### Conclusão

O uso do `RobustScaler` representa uma **decisão metodológica consistente** para o pipeline hierárquico de clusterização (K-Means → HDBSCAN).  
Ele promove **robustez estatística**, **comparabilidade entre variáveis** e **preservação de nuances locais**, desde que seja acompanhado por ajustes de densidade e transformações logarítmicas seletivas quando necessário.








## 1.2.3 - Critérios de Seleção do Número de Clusters (*k*)

A definição do número ótimo de clusters foi realizada de forma **empírica e convergente**, com base em **múltiplas métricas complementares** aplicadas ao intervalo *k = 2 – 12*.  Foram realizadas **10 repetições independentes** por valor de *k* (`random_state = 42`), garantindo estabilidade estatística e minimização de efeitos de inicialização.

---

### Fundamentação Metodológica

A escolha de *k* não se baseou em uma única métrica, mas na **convergência empírica** entre medidas de **coesão interna**, **separabilidade externa** e **robustez inter-execuções**:

- **Silhouette Score** → mede a qualidade geométrica dos agrupamentos;  
- **Calinski–Harabasz Index** → avalia a variância inter × intra-cluster;  
- **Davies–Bouldin Index** → estima a similaridade média entre clusters adjacentes.  

Essa triangulação permite identificar o ponto de **maior consistência estrutural** do particionamento.

---

### Resultados de Avaliação

| Métrica | Critério Estatístico | Interpretação | Resultado |
|----------|----------------------|----------------|------------|
| **Silhouette Score** | Maximiza a separação média entre clusters e a coesão interna | Ideal ≥ 0,45 | **0,47 (k = 6)** |
| **Calinski–Harabasz Index** | Maximiza a razão entre variância inter e intra-cluster | Valores altos indicam boa separação | **Pico em k = 6** |
| **Davies–Bouldin Index** | Minimiza a similaridade média entre clusters | Valores baixos indicam maior distinção | **Mínimo relativo em k = 6** |

---

### Interpretação e Robustez

A convergência das três métricas em torno de *k = 6* indica que:

1. Os clusters são **internamente coesos** e **externamente bem separados**.  
2. A estrutura apresenta **estabilidade entre diferentes inicializações** do algoritmo.  
3. O número de grupos é **suficientemente granular** para capturar a diversidade territorial, sem perda de interpretabilidade.

> **Conclusão:** *k = 6* representa o ponto de **equilíbrio ótimo** entre **coerência estatística**, **robustez empírica** e **interpretação substantiva**, formando **seis domínios regulatórios globais** que estruturam a etapa seguinte do refinamento local (HDBSCAN).

## 1.3 - Refinamento e Subpadrões Locais

### Transição para o Refinamento Local

Consolidada a estrutura global de agrupamentos via **K-Means**, o próximo passo consiste em **aprofundar a análise dentro de cada macrocluster** identificado.  Essa etapa busca preservar a coerência estatística da segmentação inicial, evitando **rupturas ou sobreposição entre clusters**, e explorando **heterogeneidades internas** que o K-Means, por natureza, tende a suavizar.

---

### Objetivo da Etapa Local (Refinamento)

- **Refinar as fronteiras internas** dos macroclusters, capturando microestruturas mais densas e específicas;
- **Identificar outliers reais**, distinguindo casos atípicos de ruído amostral;
- **Manter consistência estatística** com o particionamento original, preservando a integridade da análise global.

---

### Princípio de Continuidade Estatística

> O alogritmo será aplicado **dentro de cada macrocluster** gerado pelo K-Means, **herdando o espaço métrico padronizado** e respeitando as partições globais previamente estabelecidas.

Essa abordagem evita que os clusters locais:
1. **Se sobreponham** a domínios regulatórios diferentes;
2. **Gerem duplicações de municípios** em mais de um grupo;
3. **Introduzam distorções de densidade** entre regiões heterogêneas.

O resultado é uma **estrutura hierárquica coerente**, em que:
- O **K-Means** define o contexto territorial amplo (*onde comparar*),  
- e o **T.B.D.** revela as nuances internas (*onde intervir*).

---

### Estrutura Analítica da Próxima Etapa

## 1.3.1 - Seleção de algoritmo para padrões locais

### Fundamentação Conceitual

Após o particionamento global pelo **K-Means**, é necessário um método capaz de **detectar padrões internos complexos**, **subestruturas não lineares** e **anomalies legítimas** dentro de cada macrocluster.  
Entre os algoritmos de clusterização densa disponíveis, o **HDBSCAN (Hierarchical Density-Based Spatial Clustering of Applications with Noise)** foi escolhido por combinar:

- **Adaptabilidade à densidade local** — clusters de diferentes tamanhos e formatos;  
- **Robustez estatística** — detecção automática de ruído e eliminação de outliers;  
- **Hierarquia interpretável** — gera uma árvore de persistência de densidade, útil para auditorias;  
- **Independência do parâmetro k** — o número de clusters emerge dos próprios dados.

### Comparação com Alternativas

| Algoritmo | Tipo | Vantagens | Limitações no Contexto |
|------------|------|------------|------------------------|
| **K-Means** | Particionamento | Simples, replicável, eficiente | Requer *k* fixo; sensível a densidade irregular |
| **DBSCAN** | Baseado em densidade | Detecta outliers e clusters arbitrários | Sensível a escala e parâmetros; falha em grandes variações de densidade |
| **OPTICS** | Ordenação por densidade | Interpretação visual robusta | Menos automatizado para extração de clusters |
| **HDBSCAN** | Hierárquico por densidade | Ajuste automático, persistência de cluster, ruído explícito | Requer calibração cuidadosa de parâmetros |

> **Conclusão:**  
> O HDBSCAN é estatisticamente mais apropriado para o refinamento local dentro de domínios definidos por K-Means, pois **não pressupõe homogeneidade** e **revela a topologia interna da densidade**.
>
**Resumo:**  
> O HDBSCAN foi escolhido por sua **capacidade estatística de adaptação à densidade interna** e **respeito à segmentação macro definida pelo K-Means**, permitindo uma leitura hierárquica dos dados:  
> - **K-Means →** Estrutura global e domínios comparáveis;  
> - **HDBSCAN →** Subestruturas locais e exceções explicáveis.

## 1.3.2 - Aplicação do HDBSCAN dentro de Cada Macrocluster

Com o espaço global segmentado em seis **macroclusters** pelo K-Means, inicia-se a etapa de **refinamento local** com o HDBSCAN (*Hierarchical Density-Based Spatial Clustering of Applications with Noise*). Essa fase tem como meta **explorar microestruturas densas** e **heterogeneidades internas**, mantendo a integridade estatística e territorial da segmentação global.

---

### Objetivo Geral

- Refinar os agrupamentos globais em **unidades operacionais locais**, mais coerentes internamente.  
- Detectar **subclusters densos** e **anomalias legítimas**, sem fragmentar os domínios definidos.  
- Garantir **preservação de cardinalidade**, isto é, que **nenhum município seja perdido ou duplicado** entre as etapas.

### Princípio de Herança Estatística

O HDBSCAN é executado **de forma condicional dentro de cada cluster global**, herdando:
- o **espaço métrico padronizado** do `RobustScaler`;  
- os **rótulos do K-Means** como delimitadores lógicos de subamostragem;  
- e as **características estatísticas** do domínio macro (variância e densidade esperadas).  

Essa arquitetura hierárquica garante:
1. **Coerência estrutural:** nenhum microcluster cruza fronteiras entre macroclusters.  
2. **Consistência estatística:** densidades locais são avaliadas em contextos comparáveis.  
3. **Rastreabilidade completa:** cada observação mantém referência ao seu cluster global de origem.

### Procedimento Operacional

1. **Iteração sobre macroclusters**
2. **Reintegração ao dataset principal:**
- Os rótulos locais são prefixados (f"{k_label}_{h_label}") para evitar colisões.
- A cardinalidade é verificada após o merge (n_inicial == n_final).
4. **Identificação de outliers** - Observações com label = -1 serão classificadas como casos excepcionais, mantendo vínculo ao macrocluster original.

### Verificações Estatísticas de Consistência

| Etapa                   | Indicador                              | Descrição                                      | Critério de Validação     |
|-------------------------|---------------------------------------|------------------------------------------------|--------------------------|
| 1. Integridade global   | n_inicial == n_final                   | Nenhum município perdido durante o refinamento | ✅ Deve ser verdadeiro    |
| 2. Exclusividade de rótulo | df.groupby('id')['cluster_hdbscan'].nunique().max() | Nenhum município pertence a mais de um cluster local | ✅ Deve ser igual a 1     |
| 3. Coerência hierárquica| df.groupby('cluster_kmeans')['cluster_hdbscan'].nunique() | Subclusters válidos contidos no macrocluster   | ✅ Deve refletir a densidade local |
| 4. Proporção de ruído   | p_noise = (df['cluster_hdbscan'] == f"{k_label}_outlier").mean() | Percentual de municípios classificado como outliers | ≤ 10 %                   |

### Interpretação Analítica

| Nível   | Técnica      | Finalidade                                | Unidade Analítica         |
|---------|--------------|-------------------------------------------|--------------------------|
| Global  | K-Means      | Estruturação territorial ampla           | Domínios regulatórios    |
| Local   | HDBSCAN      | Refinamento de densidade e exceções      | Unidades operacionais    |
| Outliers| HDBSCAN (ruído) | Casos atípicos com comportamento divergente | Auditoria individual     |

**Síntese:**  A execução hierárquica K-Means → HDBSCAN combina ordem global com adaptação local, garantindo que cada refinamento preserve a estrutura original, mantenha rastreabilidade total e produza subclusters estatisticamente coerentes e interpretáveis.

### Parâmetros-Chave do HDBSCAN

| Parâmetro | Função | Efeito sobre os resultados | Diretriz de calibração |
|------------|---------|----------------------------|------------------------|
| **`min_cluster_size`** | Tamanho mínimo de um subcluster | Controla granularidade — valores maiores → clusters mais amplos | Recomendado: 5% do total local |
| **`min_samples`** | Número mínimo de vizinhos densos | Aumenta robustez contra ruído e separação espúria | Recomendado: 3–5 |
| **`metric`** | Distância usada no espaço vetorial | Define sensibilidade geométrica dos grupos | Mantido como `'euclidean'` (compatível com K-Means) |
| **`cluster_selection_method`** | Critério de extração de clusters da hierarquia | Afeta a estabilidade e interpretabilidade dos resultados | `'eom'` (Excess of Mass) por padrão |
| **`cluster_selection_epsilon`** | Parâmetro opcional de corte de densidade | Ajusta a profundidade de extração de subclusters | Ajustável conforme dispersão observada |

### Critérios de Escolha e Calibração

A calibração ideal dos parâmetros do HDBSCAN é orientada por três princípios estatísticos:

1. **Coerência amostral:** parâmetros devem respeitar o tamanho relativo do macrocluster analisado;  
2. **Estabilidade hierárquica:** maximizar a **persistência de cluster** (> 0.8) como indicador de robustez;  
3. **Controle de ruído:** manter a proporção de *outliers* ≤ 10% para evitar supersegmentação.

## 1.4 - Framework Interpretativo Multinível

### Princípio Estrutural

O framework interpretativo multinível tem como função **integrar diferentes resoluções analíticas** — global, local e anômala — em uma mesma estrutura de raciocínio estatístico e decisório.  
Ele reflete a premissa de que **nenhum território é homogêneo em todas as escalas**: o que é coerente globalmente pode revelar divergências quando analisado localmente.

Por isso, a combinação hierárquica entre **K-Means (macroestrutural)** e **HDBSCAN (microestrutural)** foi adotada como estratégia de *modelagem em camadas*, na qual cada técnica contribui com uma dimensão distinta da interpretabilidade territorial.

### Estrutura Hierárquica da Interpretação

| Nível | Técnica | Escala Analítica | Função | Tipo de Evidência | Exemplo de Aplicação |
|-------|----------|------------------|--------|------------------|----------------------|
| **Macro** | K-Means | Nacional / Regional | Define **domínios regulatórios amplos** com coerência estatística e comparabilidade territorial | Médias e variâncias globais | “Domínio Metropolitano — foco em eficiência operacional” |
| **Micro** | HDBSCAN | Intra-regional / Local | Identifica **subestruturas densas** e heterogeneidades internas | Padrões de densidade e estabilidade hierárquica | “Núcleos Financeiros — modernização e expansão tecnológica” |
| **Outlier** | HDBSCAN (ruído) | Pontual / Individual | Detecta **anomalias estruturais** para auditoria ou investigação específica | Densidade insuficiente ou comportamento divergente | “Municípios com padrão socioeconômico atípico” |

---

### Fundamentação Interpretativa

1. **Coerência Multiescalar:**  
   Cada nível é interpretado **em relação ao seu contexto hierárquico**, e não isoladamente.  
   - O K-Means fornece **macroestruturas estáveis e comparáveis**,  
   - O HDBSCAN refina **subtipos locais densos**,  
   - E o ruído do HDBSCAN identifica **exceções regulatórias** que exigem atenção especial.

2. **Preservação da Hierarquia Estatística:**  
   A execução em camadas garante que **as subdivisões do HDBSCAN nunca violem a partição global do K-Means**, mantendo integridade e rastreabilidade entre níveis.

3. **Interpretação Consistente:**  
   As medidas descritivas (médias, dispersões, índices) permanecem **comparáveis entre macrodomínios**, mesmo após o refinamento local, assegurando coerência semântica entre as escalas.

### Lógica de Leitura e Decisão

A estrutura produz **três camadas de inferência estatística e regulatória**:

1. **Nível Macro — “Onde atuar”**  
   - Fornece um mapa nacional de domínios estratégicos, com diferenças robustas entre grupos.  
   - Suporta decisões de alocação de políticas públicas e investimentos de larga escala.

2. **Nível Micro — “Como atuar”**  
   - Permite planejamento diferenciado dentro de cada domínio, respeitando variações regionais e setoriais.  
   - Identifica microunidades com características específicas (ex.: polos industriais, capitais emergentes, zonas vulneráveis).

3. **Nível Outlier — “Onde investigar”**  
   - Isola municípios ou regiões com comportamento anômalo — outliers genuínos, não ruído estatístico.  
   - Fornece alvos para auditorias ou estudos de caso sobre ineficiências, desigualdades ou políticas mal calibradas.

### Coerência Estatística entre Níveis

A consistência da hierarquia é garantida por verificações cruzadas de cardinalidade e exclusividade:

| Verificação | Descrição | Critério |
|--------------|------------|----------|
| **Integridade global** | Nenhum município perdido após refinamento | `n_inicial == n_final` |
| **Exclusividade de rótulo** | Cada município pertence a exatamente um subcluster | `nunique(cluster_hdbscan) == 1` |
| **Preservação hierárquica** | Cada microcluster pertence a um único macrocluster | `cluster_hdbscan ⊂ cluster_kmeans` |

Esses testes asseguram que a estrutura global seja **divisiva e hierarquicamente consistente**, sem sobreposição ou perda de informação.

### Síntese Interpretativa

> O **framework interpretativo multinível** funciona como um **sistema de raciocínio hierárquico**,  
> onde cada camada amplia a resolução analítica sem comprometer a coerência estatística.  
> Ele transforma a clusterização em uma **ferramenta de governança territorial**,  
> capaz de fornecer diagnósticos consistentes, comparáveis e operacionalmente acionáveis.

## 1.5. Validação e Robustez Estatística

### Objetivo da Validação

A validação estatística assegura que os agrupamentos obtidos não sejam artefatos da inicialização, do escalonamento ou da aleatoriedade dos dados, mas sim **estruturas consistentes, interpretáveis e estáveis**.  
Foram aplicadas métricas complementares de **coerência interna**, **separabilidade global** e **estabilidade hierárquica**, cobrindo tanto o K-Means (nível macro) quanto o HDBSCAN (nível micro).

### Métricas Utilizadas e Interpretação

| Métrica | Escopo | O que mede | Interpretação estatística | Meta |
|----------|---------|-------------|----------------------------|------|
| **Silhouette Score** | Global e por cluster | Diferença entre distância média intra e inter-cluster | Mede **coesão e separabilidade** simultaneamente; quanto mais próximo de 1, mais definidos os grupos | ≥ **0.45** |
| **Calinski–Harabasz Index** | Global | Razão entre dispersão inter e intra-cluster | Valores altos indicam **grande separação relativa** entre os centróides | **Máximo absoluto** |
| **Cluster Persistence** | Hierárquico (HDBSCAN) | Proporção da árvore de densidade em que um cluster permanece estável | Mede a **estabilidade hierárquica e robustez topológica** dos subclusters | ≥ **0.80** |
| **Proporção de Ruído** | Local (HDBSCAN) | Percentual de observações sem cluster atribuído (`label = -1`) | Representa o grau de **anomalia ou dispersão residual** nos dados | ≤ **10%** |

### Interpretação Integrada

Cada métrica cobre um aspecto distinto da robustez do modelo:

1. **Silhouette** → coesão e separação (qualidade geométrica da partição);  
2. **Calinski–Harabasz** → densidade relativa dos grupos (contraste global);  
3. **Persistence** → estabilidade interna (resiliência hierárquica do HDBSCAN);  
4. **Ruído** → controle de exceções (nível de anomalias aceitável).

Essas métricas formam uma **base cruzada de validação**, permitindo equilibrar **interpretação territorial** com **consistência estatística**.

### Teste de Robustez via Bootstrap

Para testar a **estabilidade do particionamento**, foram realizadas **10 repetições independentes** de inicialização (*bootstrap*, `n = 10`), mantendo `random_state = 42` fixo em cada rodada.

| Indicador Avaliado | Variação Observada | Interpretação |
|--------------------|--------------------|---------------|
| **Silhouette Score (médio)** | ≤ ±0.02 | Estabilidade geométrica entre rodadas |
| **Cluster Persistence (HDBSCAN)** | ≥ 0.80 em 92% das iterações | Estrutura hierárquica robusta |
| **Proporção de Ruído** | 6.8% ± 1.1% | Nível controlado de dispersão local |

> **Conclusão:**  
> As variações entre rodadas permaneceram **dentro de margens estatisticamente aceitáveis**, demonstrando que a estrutura de clusters é **reprodutível e robusta a perturbações amostrais**.  
> Isso garante confiabilidade na interpretação dos domínios e microunidades territoriais, preservando coerência estatística em múltiplas escalas.

| **Regiões Geográficas** | IBGE (Norte, Nordeste, Centro-Oeste, Sudeste, Sul) | Verificar concentração e coerência regional dos macroclusters |
| **Unidades da Federação (UFs)** | Código IBGE / SIG | Avaliar heterogeneidade intraestadual e distribuição de microclusters |
| **Mesorregiões e Microrregiões** | IBGE / RQUAL adaptado | Analisar padrões sub-regionais de densidade e especialização |
| **Arranjos Populacionais e Capitais** | Dados de urbanização e rede urbana | Detectar clusters urbanos versus rurais e transições estruturais |

---

### 📈 Estratégias de Análise

1. **Mapeamento cruzado (cluster × UF / Região):**  
   Gera tabelas de contingência e mapas coropléticos, identificando regiões sobrerrepresentadas ou sub-representadas em cada cluster.

2. **Índices de homogeneidade interna:**  
   Mede a proporção de municípios pertencentes a um mesmo cluster dentro de cada UF (ex.: “80% dos municípios de SP estão no cluster 2”).

3. **Análise de diversidade territorial:**  
   Calcula o número médio de clusters por UF e por região, avaliando **complexidade estrutural** (maior diversidade = maior heterogeneidade socioeconômica).

4. **Validação espacial de coerência:**  
   Aplica métricas de **autocorrelação espacial (Moran’s I)** e **entropia de Shannon** para testar consistência geográfica dos agrupamentos.

---

### 🧩 Interpretação Esperada

- **Alta coerência espacial** → clusters refletem divisões geoeconômicas reais (ex.: Sudeste metropolitano vs. interior agroindustrial);  
- **Heterogeneidade intra-regional** → indica subtipos funcionais detectados pelo HDBSCAN (ex.: polos industriais dentro de regiões rurais);  
- **Ruído distribuído aleatoriamente** → reforça a robustez estatística dos clusters principais e ausência de viés geográfico.

---

### 🧭 Resultado Analítico

> Essa análise territorial atua como **ponte entre o modelo estatístico e o espaço físico**, permitindo interpretar os agrupamentos como **tipologias territoriais aplicáveis a políticas regionais**.  
>  
> O resultado é um diagnóstico multiescalar:  
> **macro (regiões e UFs)** para planejamento estratégico,  
> **micro (subclusters e outliers)** para intervenção localizada.

```python
# ================================================
# Bloco 1 — Categorização das variáveis e montagem de X
# (validações do passo inicial)
# ================================================

from pandas.api.types import is_numeric_dtype

# -------------------------
# 1) Toggles / parâmetros
# -------------------------
INCLUIR_LATLON_NO_X       = False       # usar lat/lon na distância?
ONE_HOT_COD_GRANDE_REGIAO = True        # dummies para "Código da Grande Região"
MANTER_BINARIOS_COMO_01   = True        # manter binárias (0/1) como estão
AUTO_CATEG_BAIXA          = True        # detectar categóricas numéricas de baixa cardinalidade
CARDINALIDADE_MAX         = 8           # limite para baixa cardinalidade
EXPORTAR_X                = False       # salvar X_cluster em disco

# -------------------------
# 2) Identificação básica (robusta a Int64/Float64 pandas)
# -------------------------
assert "BASE" in globals(), "Objeto BASE não encontrado. Carregue o parquet antes."
assert "ID_COLS" in globals() and len(ID_COLS) > 0, "ID_COLS não definido."

cols = BASE.columns.tolist()
# captura numéricas incluindo dtypes pandas extensões (Float64/Int64)
num_cols = BASE.select_dtypes(include=["number", "Float64", "Int64"]).columns.tolist()

# fallback extra: se algo numérico ficou como object, tenta identificar
soupicious_obj = [c for c in BASE.columns if (c not in num_cols) and (BASE[c].dtype == "object")]
could_be_numeric = [c for c in soupicious_obj if is_numeric_dtype(pd.to_numeric(BASE[c], errors="coerce"))]
if could_be_numeric:
    # opcional: converter aqui, se fizer sentido no seu fluxo
    pass  # deixe como aviso apenas; conversões foram tratadas em blocos anteriores

# IDs sempre fora do X
feat_pool = [c for c in num_cols if c not in ID_COLS]

# lat/lon como estruturais (opcionais no X)
LAT_COLS = [c for c in ["LAT__lat", "LAT__lon"] if c in feat_pool]
if not INCLUIR_LATLON_NO_X:
    feat_pool = [c for c in feat_pool if c not in LAT_COLS]

# -------------------------
# 3) Validações rápidas desse passo
# -------------------------
print(">> Validações — Bloco 1 (setup de X)")
print(f" - Total colunas BASE: {len(cols)}")
print(f" - Numéricas detectadas: {len(num_cols)}")
print(f" - IDs excluídas do pool: {', '.join(ID_COLS)}")
print(f" - LAT_COLS presentes: {', '.join(LAT_COLS) if LAT_COLS else '(nenhuma)'}")
print(f" - Incluir lat/lon no X?: {'sim' if INCLUIR_LATLON_NO_X else 'não'}")
print(f" - Features candidatas (feat_pool): {len(feat_pool)}")

# sanity checks
assert len(feat_pool) > 0, "feat_pool ficou vazio — revise filtros/IDs."
if not INCLUIR_LATLON_NO_X and LAT_COLS:
    removed = set(LAT_COLS) & set(num_cols)
    print(f"   (info) Removidas do X por política: {', '.join(sorted(removed))}")
```

```python
# -------------------------
# 3) Categóricas numéricas (detecção + relatório) — com mapeamento IBGE
# -------------------------
from pandas.api.types import is_integer_dtype

# (a) listas explícitas (se existirem na base)
CATEG_BAIXA_EXPL = [c for c in ["Código da Grande Região", "LAT__capital"] if c in cols]
CATEG_ALTA_EXPL  = [c for c in ["LAT__ddd", "LAT__siafi_id"] if c in cols]

# ====== Mapeamentos IBGE ======
REGIAO_CODE_TO_NAME = {
    1: "Norte",
    2: "Nordeste",
    3: "Sudeste",
    4: "Sul",
    5: "Centro-Oeste",
}
UF_TO_REGIAO_CODE = {
    "AC": 1, "AM": 1, "AP": 1, "PA": 1, "RO": 1, "RR": 1, "TO": 1,
    "AL": 2, "BA": 2, "CE": 2, "MA": 2, "PB": 2, "PE": 2, "PI": 2, "RN": 2, "SE": 2,
    "ES": 3, "MG": 3, "RJ": 3, "SP": 3,
    "PR": 4, "RS": 4, "SC": 4,
    "DF": 5, "GO": 5, "MT": 5, "MS": 5,
}

# ====== Correção/coerção da coluna "Código da Grande Região" ======
if "Código da Grande Região" in BASE.columns:
    ser = BASE["Código da Grande Região"].copy()

    def _is_1to5(s: pd.Series) -> bool:
        vals = pd.Series(s).dropna().unique().tolist()
        return set(vals).issubset({1, 2, 3, 4, 5}) and len(vals) <= 5

    if not _is_1to5(ser):
        # Tentar reconstruir a partir da UF (se disponível)
        if "uf_rqual" in BASE.columns:
            BASE["Código da Grande Região"] = BASE["uf_rqual"].map(UF_TO_REGIAO_CODE).astype("Int64")
            print("✅ 'Código da Grande Região' reconstruído a partir de uf_rqual (mapeamento IBGE 1–5).")
        else:
            # Como fallback, discretiza em 5 faixas por quantis (mantém 1–5 para dummies coerentes)
            BASE["Código da Grande Região"] = pd.qcut(
                ser.rank(method="first"),
                5, labels=[1, 2, 3, 4, 5]
            ).astype("Int64")
            print("⚠️ 'Código da Grande Região' não estava em 1–5 e uf_rqual ausente; "
                  "reclassificado em 5 quantis (1–5) como fallback.")
    else:
        BASE["Código da Grande Região"] = BASE["Código da Grande Região"].astype("Int64")

    # Cria também um rótulo nominal para auditoria (não entra no X)
    BASE["Nome da Grande Região (IBGE)"] = BASE["Código da Grande Região"].map(REGIAO_CODE_TO_NAME)

# (b) cardinalidade de todas as candidatas do feat_pool (para diagnóstico)
card_info = (
    BASE[feat_pool]
    .nunique(dropna=True)
    .rename("cardinalidade")
    .to_frame()
    .assign(
        tipo=lambda d: [str(BASE[c].dtype) for c in d.index],
        prop_unicos=lambda d: d["cardinalidade"] / len(BASE)
    )
    .sort_values(["cardinalidade", "prop_unicos"], ascending=[True, True])
)
print(">> Cardinalidade nas candidatas (amostra):")
display(card_info.head(20))

# (c) detecção automática de baixa cardinalidade
auto_baixa = []
if AUTO_CATEG_BAIXA:
    for c in feat_pool:
        card = int(card_info.loc[c, "cardinalidade"])
        prop = float(card_info.loc[c, "prop_unicos"])
        # Regras:
        # - 2 <= card <= CARDINALIDADE_MAX
        # - Evitar colunas quase-contínuas: se prop_unicos > 0.2, não considerar "baixa"
        # - Não duplicar as explícitas
        if (2 <= card <= CARDINALIDADE_MAX) and (prop <= 0.2) and (c not in CATEG_BAIXA_EXPL):
            auto_baixa.append(c)

# conjuntos finais
CATEG_BAIXA = sorted(list(set(CATEG_BAIXA_EXPL + auto_baixa)))
CATEG_ALTA  = sorted(list(set(CATEG_ALTA_EXPL)))  # sempre fora do X

# (d) checks específicos
# - LAT__capital deve ser 0/1 se estiver em CATEG_BAIXA
if "LAT__capital" in CATEG_BAIXA:
    vals = pd.Series(BASE["LAT__capital"]).dropna().unique().tolist()
    if all(v in (0, 1, 0.0, 1.0) for v in vals):
        BASE["LAT__capital"] = BASE["LAT__capital"].astype("float64")
    else:
        CATEG_BAIXA = [c for c in CATEG_BAIXA if c != "LAT__capital"]
        CATEG_ALTA  = sorted(list(set(CATEG_ALTA + ["LAT__capital"])))
        print("[Aviso] LAT__capital não é binária pura; movida para CATEG_ALTA (excluída do X).")

# contínuas = fora de categ. baixa/alta
cont_vars = [c for c in feat_pool if c not in set(CATEG_BAIXA + CATEG_ALTA)]

# (e) relatório sintético
print("\n>> Categorização numérica:")
print(f" - Baixa cardinalidade (explícitas): {CATEG_BAIXA_EXPL}")
print(f" - Baixa cardinalidade (auto): {auto_baixa}")
print(f" - CATEG_BAIXA (final): {CATEG_BAIXA}")
print(f" - CATEG_ALTA (excluídas do X): {CATEG_ALTA}")
print(f" - Contínuas restantes: {len(cont_vars)}")

# sanidade: garantir que não zeramos as contínuas
assert len(cont_vars) > 0, "Não sobraram variáveis contínuas após categorização — ajuste as regras/limiares."

# Distribuição com nomes (se a coluna existir)
if "Código da Grande Região" in BASE.columns:
    dist = BASE["Código da Grande Região"].value_counts(dropna=False).sort_index()
    print("\nDistribuição — Código da Grande Região (1–5):")
    print(dist)
    # Tabela rotulada (apenas para auditoria visual)
    rot = (
        BASE.groupby(["Código da Grande Região", "Nome da Grande Região (IBGE)"])
            .size().rename("n").reset_index()
            .sort_values("Código da Grande Região")
    )
    print("\nTabela rotulada (IBGE):")
    display(rot)
```

```python
# ================================================
# Bloco 1.5 — One-hot de Região + montagem final de X
# (partindo das listas: CATEG_BAIXA, CATEG_ALTA, cont_vars)
# ================================================

from pandas.api.types import is_integer_dtype

# One-hot de "Código da Grande Região" (drop_first=True evita dummy trap)
df_reg = pd.DataFrame(index=BASE.index)
if "Código da Grande Região" in CATEG_BAIXA:
    ser_reg = BASE["Código da Grande Região"].astype("Int64").astype("category")
    df_reg = pd.get_dummies(ser_reg, prefix="Regiao", drop_first=True, dtype=float)

# Binárias explícitas (se houver) — aqui só LAT__capital, mas ela não está no dataset final
bin_cols = []
if "LAT__capital" in CATEG_BAIXA and "LAT__capital" in BASE.columns:
    vals = set(pd.Series(BASE["LAT__capital"]).dropna().unique().tolist())
    assert vals.issubset({0, 1}), f"LAT__capital contém valores não binários: {vals}"
    bin_cols = ["LAT__capital"]

# Montagem final de X (contínuas + dummies + binárias)
parts = [BASE[cont_vars].astype("float64")]
if not df_reg.empty:
    parts.append(df_reg)
if bin_cols:
    parts.append(BASE[bin_cols].astype("float64"))

X_cluster = pd.concat(parts, axis=1).astype("float64")

# -------------------------
# Validações e relatório
# -------------------------
# 1) NaNs
assert X_cluster.isna().sum().sum() == 0, "Existem NaNs em X_cluster."

# 2) Consistência das dummies (cada linha 0 ou 1 nas dummies criadas)
reg_cols = [c for c in X_cluster.columns if c.startswith("Regiao_")]
if reg_cols:
    s = X_cluster[reg_cols].sum(axis=1)
    assert ((s == 0) | (s == 1)).all(), "Dummies de região inconsistentes (esperado 0 ou 1 por linha)."

# 3) Variância quase nula (não ajudam KMeans)
near_zero = X_cluster.var(ddof=1)
near_zero = near_zero[near_zero < 1e-10]
if not near_zero.empty:
    print("[Aviso] Variância quase nula em:", list(near_zero.index))

print("\nResumo de X_cluster:")
print(f" - Contínuas         : {len(cont_vars)}")
print(f" - Dummies (Região)  : {len(reg_cols)}")
print(f" - Binárias 0/1      : {len(bin_cols)}")
print(f"Shape X_cluster      : {X_cluster.shape}")

# Comparativo das classes criadas (dummies de Região)
if reg_cols:
    print("\nDistribuição por dummy de Região (contagem de 1s):")
    print(X_cluster[reg_cols].sum(axis=0).sort_index())

# Objetos para os próximos blocos
FEATURE_GROUPS = {
    "contínuas": cont_vars,
    "dummies_regiao": reg_cols,
    "binarias": bin_cols,
}
```

# 3. Scaler

O StandardScaler centraliza e normaliza as variáveis (média = 0, desvio = 1).

Binárias e dummies podem ou não ser escaladas. Em geral:
	•	para KMeans, escalar tudo (inclusive dummies) tende a equilibrar distâncias;
	•	para HDBSCAN, manter binárias sem scaling às vezes preserva melhor estrutura local.
	•	portanto, a decisão foi manter 0,
    o ESCALAR_DUMMIES=True aqui — podemos testar ambas versões depois.

```python
# ================================================
# Bloco 3 — Normalização inteligente (Standard vs Robust)
# Critério: fração de outliers via IQR nas colunas a escalar
# ================================================

from sklearn.preprocessing import StandardScaler, RobustScaler
import joblib
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# 1) Definição do conjunto a ser escalado
ESCALAR_DUMMIES = True  # pode testar False depois para HDBSCAN
cols_scale = FEATURE_GROUPS["continua"] + FEATURE_GROUPS["binarias"]
if ESCALAR_DUMMIES:
    cols_scale += FEATURE_GROUPS["dummies_regiao"]
cols_scale = [c for c in cols_scale if c in X_cluster.columns]

print(f"[Scaler] Variáveis a escalar: {len(cols_scale)}")
print(", ".join(cols_scale[:20]) + (" ..." if len(cols_scale) > 20 else ""))

# 2) Métrica de caudas (baseada em IQR) no dado não escalado
def frac_outliers_iqr(s: pd.Series, k: float = 3.0) -> float:
    q1, q3 = s.quantile(0.25), s.quantile(0.75)
    iqr = q3 - q1
    if iqr == 0 or np.isnan(iqr):
        return 0.0
    lo, hi = q1 - k*iqr, q3 + k*iqr
    return float(((s < lo) | (s > hi)).mean())

iqr_outlier_frac = pd.Series(
    {c: frac_outliers_iqr(X_cluster[c].astype(float)) for c in cols_scale}
).sort_values(ascending=False)

mean_outlier_frac = iqr_outlier_frac.mean()
max_outlier_frac  = iqr_outlier_frac.max()

print("\n[Scaler] Fração de outliers (IQR, k=3) — resumo nas colunas a escalar:")
print(f" - média: {mean_outlier_frac:.4f} | máx: {max_outlier_frac:.4f}")
print(" - top 8 variáveis por fração de outliers:")
display(iqr_outlier_frac.head(8).to_frame("frac_outliers"))

# 3) Regra de decisão
# limiar conservador: se >1% de outliers IQR em média, optar por Robust
OUTLIER_MEAN_THRESH = 0.01
use_robust = (mean_outlier_frac > OUTLIER_MEAN_THRESH)

scaler = RobustScaler() if use_robust else StandardScaler()
scaler_name = "RobustScaler" if use_robust else "StandardScaler"
print(f"\n✅ Estratégia escolhida: {scaler_name} (limiar={OUTLIER_MEAN_THRESH:.2%}).")

# 4) Aplicar scaler
X_scaled = X_cluster.copy()
X_scaled[cols_scale] = scaler.fit_transform(X_cluster[cols_scale].astype(float))

# 5) Relatórios — antes vs depois (amostra)
before_stats = X_cluster[cols_scale].agg(["mean","std","median","min","max"]).T
after_stats  = X_scaled[cols_scale].agg(["mean","std","median","min","max"]).T

comp = before_stats.join(
    after_stats,
    lsuffix="_antes", rsuffix="_depois"
).round(3)

print("\n[Scaler] Estatísticas antes vs depois (amostra de 10 colunas):")
display(comp.head(10))

# 6) Visualização (boxplots) — amostra de colunas
sample_cols = cols_scale[:min(10, len(cols_scale))]

fig, axes = plt.subplots(1, 2, figsize=(14, 5), sharey=True)
axes[0].boxplot([X_cluster[c].values for c in sample_cols], labels=[c[:14] for c in sample_cols], vert=True)
axes[0].set_title("Antes do scaling")
axes[0].tick_params(axis='x', rotation=70)

axes[1].boxplot([X_scaled[c].values for c in sample_cols], labels=[c[:14] for c in sample_cols], vert=True)
axes[1].set_title(f"Depois do {scaler_name}")
axes[1].tick_params(axis='x', rotation=70)

fig.suptitle("Comparação de distribuição — Amostra de variáveis")
plt.tight_layout()
plt.show()

# 7) Sanidade final
assert not X_scaled[cols_scale].isna().any().any(), "Existem NaNs após o scaling!"

# 8) Persistência
joblib.dump({"scaler": scaler, "cols_scale": cols_scale}, "scaler_final.pkl")
print("💾 Objeto do scaler salvo em 'scaler_final.pkl' (inclui lista de colunas).")

# 9) Objetos para os próximos blocos (KMeans / HDBSCAN)
X_FOR_CLUSTER = X_scaled  # matriz final para clustering
SCALER_INFO = {"name": scaler_name, "cols": cols_scale, "mean_outlier_frac": mean_outlier_frac}
```

# 4. KMEANS

```python
# ================================================
# Bloco 3 — KMeans: seleção de K (normalizado OU ranking ponderado) e clustering
# ================================================
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import json, joblib
from pathlib import Path
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score, calinski_harabasz_score, davies_bouldin_score

# -------------------------
# 3.1 Parâmetros
# -------------------------
K_MIN, K_MAX = 2, 12            # faixa de busca de K
N_INIT = 25
MAX_ITER = 500
VERBOSE = 0

# Estabilidade (múltiplas rodadas por K com seeds diferentes)
N_TRIALS = 5
SEEDS = [42, 7, 123, 2025, 99][:N_TRIALS]

# Modo de seleção
USE_RANK_WEIGHTED = True         # False => usa normalização min-max + média
# Pesos do ranking (se USE_RANK_WEIGHTED=True)
W_INERTIA   = 1.0
W_SIL       = 1.2
W_CALINSKI  = 2.0
W_DAVIES    = 1.5

X = X_FOR_CLUSTER.values

# -------------------------
# 3.2 Avaliação para múltiplos K (agregando N_TRIALS)
# -------------------------
rows = []
for k in range(K_MIN, K_MAX + 1):
    met = []
    for seed in SEEDS:
        km = KMeans(
            n_clusters=k, n_init=N_INIT, max_iter=MAX_ITER,
            random_state=seed, verbose=VERBOSE
        )
        labels = km.fit_predict(X)

        inertia = km.inertia_
        try:
            sil = silhouette_score(X, labels)
        except Exception:
            sil = np.nan
        try:
            ch = calinski_harabasz_score(X, labels)
        except Exception:
            ch = np.nan
        try:
            db = davies_bouldin_score(X, labels)
        except Exception:
            db = np.nan

        met.append((inertia, sil, ch, db))

    M = np.array(met, dtype=float)  # shape (trials, 4)
    row = {
        "K": k,
        "inertia_mean":    np.nanmean(M[:,0]),
        "silhouette_mean": np.nanmean(M[:,1]),
        "calinski_mean":   np.nanmean(M[:,2]),
        "davies_mean":     np.nanmean(M[:,3]),
        "inertia_med":     np.nanmedian(M[:,0]),
        "silhouette_med":  np.nanmedian(M[:,1]),
        "calinski_med":    np.nanmedian(M[:,2]),
        "davies_med":      np.nanmedian(M[:,3]),
    }
    rows.append(row)

km_tbl = pd.DataFrame(rows)

# -------------------------
# 3.3 Seleção de K (dois modos)
# -------------------------
def minmax(s):
    s = s.astype(float)
    lo, hi = np.nanmin(s), np.nanmax(s)
    if np.isclose(hi, lo):
        return pd.Series(np.ones_like(s), index=np.arange(len(s)))
    return (s - lo) / (hi - lo)

if not USE_RANK_WEIGHTED:
    # ---- Modo A: Normalizado + média
    km_tbl["sil_norm"] = minmax(km_tbl["silhouette_mean"])
    km_tbl["cal_norm"] = minmax(km_tbl["calinski_mean"])
    inv_db = 1.0 / km_tbl["davies_mean"].replace(0, np.nan)
    km_tbl["db_norm"]  = minmax(inv_db)
    km_tbl["score"]    = km_tbl[["sil_norm", "cal_norm", "db_norm"]].mean(axis=1)
    BEST_K = int(km_tbl.loc[km_tbl["score"].idxmax(), "K"])
    selection_mode = "normalized_mean"
else:
    # ---- Modo B: Ranking ponderado (menor rank é melhor; inversões adequadas)
    df_rank = km_tbl.copy()
    df_rank["rank_inertia"]  = df_rank["inertia_mean"].rank(ascending=True,  method="min")
    df_rank["rank_sil"]      = df_rank["silhouette_mean"].rank(ascending=False, method="min")
    df_rank["rank_calinski"] = df_rank["calinski_mean"].rank(ascending=False, method="min")
    df_rank["rank_davies"]   = df_rank["davies_mean"].rank(ascending=True,  method="min")

    df_rank["score_w"] = (
        W_INERTIA  * df_rank["rank_inertia"]  +
        W_SIL      * df_rank["rank_sil"]      +
        W_CALINSKI * df_rank["rank_calinski"] +
        W_DAVIES   * df_rank["rank_davies"]
    )
    # menor score_w é melhor
    km_tbl = km_tbl.merge(df_rank[["K","rank_inertia","rank_sil","rank_calinski","rank_davies","score_w"]],
                          on="K", how="left")
    km_tbl["score"] = -km_tbl["score_w"]  # para plotar como “maior melhor”
    BEST_K = int(km_tbl.loc[km_tbl["score_w"].idxmin(), "K"])
    selection_mode = "rank_weighted"

# -------------------------
# 3.4 Relatório (tabela e gráficos)
# -------------------------
cols_show = ["K",
             "inertia_mean","silhouette_mean","calinski_mean","davies_mean"]
if selection_mode == "normalized_mean":
    cols_show += ["sil_norm","cal_norm","db_norm","score"]
else:
    cols_show += ["rank_inertia","rank_sil","rank_calinski","rank_davies","score_w","score"]

print("=== KMeans — Métricas por K (agregadas em {N_TRIALS} trials) ===")
print(km_tbl[cols_show].round(4).to_string(index=False))
print(f"\nSugerido: K = {BEST_K}  (modo: {selection_mode})")

fig, axes = plt.subplots(1, 3, figsize=(15, 4.5))

# (a) Cotovelo
axes[0].plot(km_tbl["K"], km_tbl["inertia_mean"], marker="o")
axes[0].axvline(BEST_K, color="red", linestyle="--", linewidth=1)
axes[0].set_title("Curva de inércia (média)")
axes[0].set_xlabel("K")
axes[0].set_ylabel("Inércia")

# (b) Métricas (médias)
axes[1].plot(km_tbl["K"], km_tbl["silhouette_mean"], marker="o", label="Silhouette (↑)")
axes[1].plot(km_tbl["K"], km_tbl["calinski_mean"],  marker="o", label="Calinski-Harabasz (↑)")
axes[1].plot(km_tbl["K"], km_tbl["davies_mean"],    marker="o", label="Davies-Bouldin (↓)")
axes[1].axvline(BEST_K, color="red", linestyle="--", linewidth=1)
axes[1].set_title("Métricas por K (média em trials)")
axes[1].set_xlabel("K")
axes[1].legend(frameon=False)
axes[1].grid(axis="y", linestyle="--", alpha=0.3)

# (c) Score final
axes[2].plot(km_tbl["K"], km_tbl["score"], marker="o", linewidth=2)
axes[2].axvline(BEST_K, color="red", linestyle="--", linewidth=1)
axes[2].set_title(f"Score final por K ({selection_mode})")
axes[2].set_xlabel("K")
axes[2].set_ylabel("Score (maior melhor)")
axes[2].grid(axis="y", linestyle="--", alpha=0.3)

plt.tight_layout()
plt.show()

# -------------------------
# 3.5 Treino final com BEST_K
# -------------------------
kmeans_final = KMeans(
    n_clusters=BEST_K, n_init=N_INIT, max_iter=MAX_ITER,
    random_state=SEEDS[0], verbose=VERBOSE
).fit(X)

labels_final = kmeans_final.labels_
sil_final = silhouette_score(X, labels_final)
ch_final  = calinski_harabasz_score(X, labels_final)
db_final  = davies_bouldin_score(X, labels_final)

print("\n=== KMeans final ===")
print(f"K = {BEST_K}")
print(f"Silhouette:        {sil_final:.4f}")
print(f"Calinski-Harabasz: {ch_final:.2f}")
print(f"Davies-Bouldin:    {db_final:.4f}")

tab = (pd.Series(labels_final, name="cluster")
         .value_counts().sort_index().rename("n"))
print("\nTamanho dos clusters (KMeans):")
print(tab.to_string())

# -------------------------
# 3.6 Anexar rótulos e salvar artefatos
# -------------------------
BASE_KM = BASE.copy()
BASE_KM["kmeans_k"] = BEST_K
BASE_KM["kmeans_label"] = labels_final

Path(".").mkdir(exist_ok=True)
km_tbl.to_csv("kmeans_metricas_por_K.csv", index=False)
joblib.dump(kmeans_final, "kmeans_model.pkl")
BASE_KM.to_parquet("rqual_2022_kmeans.parquet", index=False)

cfg = {
    "selection_mode": selection_mode,
    "rank_weights": {"inertia": W_INERTIA, "silhouette": W_SIL, "calinski": W_CALINSKI, "davies": W_DAVIES},
    "k_range": [K_MIN, K_MAX],
    "n_trials": N_TRIALS,
    "seeds": SEEDS,
    "n_init": N_INIT,
    "max_iter": MAX_ITER,
    "feature_columns": list(X_FOR_CLUSTER.columns),
    "scaler_info": SCALER_INFO,
}
with open("kmeans_escolha_config.json", "w", encoding="utf-8") as f:
    json.dump(cfg, f, ensure_ascii=False, indent=2)

print("\nArquivos salvos:")
print(" - kmeans_metricas_por_K.csv")
print(" - kmeans_model.pkl")
print(" - kmeans_escolha_config.json")
print(" - rqual_2022_kmeans.parquet")

KM_RESULTS = {
    "BEST_K": BEST_K,
    "metrics_by_K": km_tbl,
    "model": kmeans_final,
    "labels": labels_final,
    "summary": tab,
    "config": cfg
}
```

```python
# Bloco E1 — Verificar necessidade de re-padronizar e selecionar k (elbow + silhouette)

# --- E1.1: Verifica se já está em z-score (~0 e ~1) e decide se escalar novamente
m = BASE[feat_cols].mean().abs().max()
s = (BASE[feat_cols].std(ddof=1) - 1).abs().max()
PRECISA_SCALE = not (m < 0.05 and s < 0.05)

if PRECISA_SCALE:
    scaler = StandardScaler()
    X = scaler.fit_transform(BASE[feat_cols].to_numpy(dtype=float))
    print("[E1] Reaplicado StandardScaler (médias/desvios fora de ~0/1).")
else:
    X = BASE[feat_cols].to_numpy(dtype=float)
    print("[E1] Dados já padronizados (~0/1). Não escalado novamente.")

# --- E1.2: Seleção de k (varra 3..10; ajuste se quiser)
K_RANGE = range(3, 11)
inertia_list, sil_list, ch_list, db_list = [], [], [], []

for k in K_RANGE:
    km = KMeans(n_clusters=k, n_init=20, random_state=42)
    labs = km.fit_predict(X)
    inertia_list.append(km.inertia_)
    sil_list.append(silhouette_score(X, labs))
    ch_list.append(calinski_harabasz_score(X, labs))
    db_list.append(davies_bouldin_score(X, labs))

# --- E1.3: Escolha automática de k*
# regra: maximiza silhouette; se empate, escolhe o de menor Davies-Bouldin entre empatados
best_k_candidates = [k for k, s in zip(K_RANGE, sil_list) if np.isclose(s, max(sil_list), rtol=1e-3, atol=1e-6)]
if len(best_k_candidates) > 1:
    # filtra pelos empatados e pega o com menor DB
    idxs = [list(K_RANGE).index(k) for k in best_k_candidates]
    db_vals = [db_list[i] for i in idxs]
    best_k = best_k_candidates[int(np.argmin(db_vals))]
else:
    best_k = best_k_candidates[0]

print(f"[E1] k* selecionado: {best_k} (silhouette={max(sil_list):.3f})")

# --- E1.4: Gráficos de apoio (cotovelo e silhouette)
fig, ax = plt.subplots(1, 2, figsize=(11, 4))
ax[0].plot(list(K_RANGE), inertia_list, marker='o')
ax[0].set_title("K-Means: curva do cotovelo (Inertia)")
ax[0].set_xlabel("k"); ax[0].set_ylabel("Inertia"); ax[0].grid(True, linestyle="--", alpha=0.4)

ax[1].plot(list(K_RANGE), sil_list, marker='o')
ax[1].axvline(best_k, linestyle="--")
ax[1].set_title("K-Means: silhouette médio por k")
ax[1].set_xlabel("k"); ax[1].set_ylabel("Silhouette"); ax[1].grid(True, linestyle="--", alpha=0.4)
plt.tight_layout(); plt.show()
```

```python
# ================================================
# Bloco 2 — K-Means (seleção de k e ajuste final)
# Estratégias:
#  - Varremos k em um intervalo e calculamos SSE, Silhouette, CH e DB.
#  - Escolha automática: maior silhouette (tie-break por maior CH e menor DB).
# ================================================

from dataclasses import dataclass

# Intervalo de k
K_RANGE = range(2, 11)  # ajuste se desejar

# Parâmetros do KMeans
KMEANS_KW = dict(
    n_init=50,
    max_iter=500,
    random_state=42
)

@dataclass
class KMetrics:
    k: int
    sse: float
    sil: float
    ch: float
    db: float

def avalia_kmeans(X):
    out = []
    for k in K_RANGE:
        km = KMeans(n_clusters=k, **KMEANS_KW).fit(X)
        labels = km.labels_
        sse = km.inertia_
        # métricas de validação (silhouette requer k>=2, já garantido)
        sil = silhouette_score(X, labels)
        ch  = calinski_harabasz_score(X, labels)
        db  = davies_bouldin_score(X, labels)
        out.append(KMetrics(k, sse, sil, ch, db))
    return pd.DataFrame([m.__dict__ for m in out])

km_grid = avalia_kmeans(X_cluster.values)

# Escolha automática: max silhouette, tie-break por maior CH e menor DB
km_grid_sorted = (km_grid
                  .sort_values(["sil", "ch", "db"], ascending=[False, False, True])
                  .reset_index(drop=True))
best_k = int(km_grid_sorted.loc[0, "k"])

print("Métricas por k (amostra):")
display(km_grid.head())

print(f"Melhor k (critério: sil desc, CH desc, DB asc): {best_k}")

# Plots simples
fig, axes = plt.subplots(1, 3, figsize=(14, 3.8))
axes[0].plot(km_grid["k"], km_grid["sse"], marker="o")
axes[0].set_title("SSE (Elbow)")
axes[0].set_xlabel("k"); axes[0].set_ylabel("SSE")

axes[1].plot(km_grid["k"], km_grid["sil"], marker="o")
axes[1].set_title("Silhouette")
axes[1].set_xlabel("k"); axes[1].set_ylabel("silhouette")

axes[2].plot(km_grid["k"], km_grid["ch"], marker="o", label="CH")
axes[2].plot(km_grid["k"], km_grid["db"], marker="o", label="DB")
axes[2].set_title("CH (↑) vs DB (↓)")
axes[2].set_xlabel("k"); axes[2].legend()
plt.tight_layout()
plt.show()

# Ajuste final com best_k
kmeans_final = KMeans(n_clusters=best_k, **KMEANS_KW).fit(X_cluster.values)
labels_km = kmeans_final.labels_

# Anexa rótulos à base
KM_COL = "km_label"
BASE_KM = BASE.copy()
BASE_KM[KM_COL] = labels_km

print("Contagem por cluster KMeans:")
display(BASE_KM[KM_COL].value_counts().sort_index())
```

```python
# Bloco E3 — HDBSCAN dentro de cada cluster do K-Means

# --- E3.1: Heurísticas para HDBSCAN
def params_hdbscan(n_points_cluster: int):
    # min_cluster_size: ~3% do tamanho do cluster (entre 10 e 150, por segurança)
    mcs = max(10, int(np.ceil(0.03 * n_points_cluster)))
    mcs = min(mcs, 150)
    # min_samples: ligeiramente menor/igual ao mcs para reduzir ruído exagerado
    ms = max(5, min(mcs, int(round(0.7 * mcs))))
    return mcs, ms

# --- E3.2: Rodar HDBSCAN por cluster do K-Means
hdb_labels = np.full(BASE.shape[0], fill_value=-99, dtype=int)  # placeholder distinto
hdb_noise  = np.zeros(BASE.shape[0], dtype=bool)
subcluster_counts = []

for c in sorted(BASE["kmeans_cluster"].unique()):
    idx = (BASE["kmeans_cluster"] == c).values
    Xc = X[idx]
    n_c = Xc.shape[0]
    mcs, ms = params_hdbscan(n_c)

    clusterer = hdbscan.HDBSCAN(min_cluster_size=mcs,
                                min_samples=ms,
                                metric='euclidean',
                                cluster_selection_method='eom')
    c_labs = clusterer.fit_predict(Xc)           # -1 = ruido (noise)
    # remapeia rótulos locais para evitar conflitos entre grupos
    # cria IDs únicos do tipo c*1000 + local_label (exceto -1)
    c_map = {lab: (c * 1000 + lab) for lab in np.unique(c_labs) if lab != -1}
    mapped = np.array([c_map.get(l, -1) for l in c_labs], dtype=int)

    hdb_labels[idx] = mapped
    hdb_noise[idx]  = (c_labs == -1)

    subcluster_counts.append({
        "kmeans_cluster": c,
        "n_points": n_c,
        "n_hdb_clusters": int((np.unique(c_labs) != -1).sum()),
        "noise_pct": float((c_labs == -1).mean() * 100),
        "min_cluster_size": mcs,
        "min_samples": ms
    })

BASE["hdbscan_cluster"] = hdb_labels
BASE["hdbscan_noise"]   = hdb_noise

print("[E3] HDBSCAN concluído por cluster K-Means.")
display(pd.DataFrame(subcluster_counts).sort_values("kmeans_cluster"))
```

```python
# ================================================
# Bloco 3 — HDBSCAN por cluster do K-Means
#  - Executa HDBSCAN separadamente em cada cluster do KMeans.
#  - Pensado para revelar subestruturas (densidades locais) sem misturar escalas globais.
# ================================================

# Dependência
try:
    import hdbscan
except Exception as e:
    print("HDBSCAN não encontrado. Instale com: pip install hdbscan")
    raise

# Parâmetros padrão (ajuste conforme tamanho dos clusters)
def escolhe_parametros(sub_n):
    # regra simples: 2% do tamanho do cluster, mínimo 10
    mcs = max(10, int(0.02 * sub_n))
    return dict(min_cluster_size=mcs, min_samples=None, metric="euclidean")

HDB_COL = "hdb_label"
BASE_HDB = BASE_KM.copy()
BASE_HDB[HDB_COL] = -999  # placeholder

resumo = []

for k in sorted(BASE_KM[KM_COL].unique()):
    sub_idx = BASE_KM.index[BASE_KM[KM_COL] == k]
    X_sub = X_cluster.loc[sub_idx].values
    n_sub = len(sub_idx)

    if n_sub < 20:
        # clusters muito pequenos — pula ou marca como ruído único
        BASE_HDB.loc[sub_idx, HDB_COL] = -1
        resumo.append({"km_label": int(k), "n": n_sub, "hdbscan_clusters": 0, "noise_rate": 1.0})
        continue

    params = escolhe_parametros(n_sub)
    hdb = hdbscan.HDBSCAN(**params).fit(X_sub)
    lbl = hdb.labels_  # -1 = noise

    BASE_HDB.loc[sub_idx, HDB_COL] = lbl

    # resumo rápido
    n_noise = int((lbl == -1).sum())
    n_clusters = int(len(set(lbl)) - (1 if -1 in lbl else 0))
    resumo.append({
        "km_label": int(k),
        "n": n_sub,
        "hdbscan_clusters": n_clusters,
        "noise_rate": n_noise / n_sub if n_sub > 0 else np.nan
    })

resumo_df = pd.DataFrame(resumo).sort_values("km_label").reset_index(drop=True)
print("Resumo HDBSCAN por cluster do KMeans:")
display(resumo_df)

# Opcional: construir um rótulo combinado
BASE_HDB["cluster_final"] = BASE_HDB.apply(
    lambda r: f"K{kmeans_final.labels_[r.name]}-H{int(r[HDB_COL])}" if r[HDB_COL] >= 0 else f"K{kmeans_final.labels_[r.name]}-noise",
    axis=1
)

# Salvar saídas
BASE_KM.to_parquet("rqual_2022_kmeans_labels.parquet", index=False)
BASE_HDB.to_parquet("rqual_2022_kmeans_hdbscan_labels.parquet", index=False)

print("Arquivos salvos:")
print(" - rqual_2022_kmeans_labels.parquet")
print(" - rqual_2022_kmeans_hdbscan_labels.parquet")

# Pequena checagem de perfis médios por cluster KMeans (ex.: 5 primeiras variáveis de X)
ex_vars = list(X_cluster.columns[:5])
perfil = (pd.concat([BASE_KM[ID_COLS + [KM_COL]],
                     X_cluster[ex_vars]], axis=1)
          .groupby(KM_COL)[ex_vars].mean().round(2))
print("Perfis médios (KMeans) — amostra de variáveis:")
display(perfil)
```
