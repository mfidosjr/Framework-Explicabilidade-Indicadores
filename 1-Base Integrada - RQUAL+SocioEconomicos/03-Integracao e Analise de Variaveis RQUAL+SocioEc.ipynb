```python
# 1. Leitura das bases

import pandas as pd
from pathlib import Path

ARQ_RQUAL = Path("rqual_2022_consolidado_clean.parquet")
ARQ_SOCIO = Path("base_socioeconomica_completa.xlsx")

rqual = pd.read_parquet(ARQ_RQUAL)
soc = pd.read_excel(ARQ_SOCIO, engine="calamine")
```

```python
# 2. Normalização das chaves

rqual["cod_mun"] = rqual["cod_mun"].astype(int)
soc["cod_mun"] = pd.to_numeric(soc["cod_mun"], errors="coerce").astype("Int64")
```

```python
# 3. Merge das bases

df_full = soc.merge(rqual, on="cod_mun", how="left", suffixes=("_soc","_rqual"))
print(f"[Base Integrada] {len(df_full):,} municípios | nulos totais: {df_full.isna().sum().sum()}")
```

```python
#4. Verificação das colunas

print("\nVariáveis RQUAL:")
print([c for c in df_full.columns if c.startswith("IND") or c.startswith("INF")])

print("\nVariáveis socioeconômicas:")
print([c for c in df_full.columns if c not in ["cod_mun","uf","nome_mun","lat","lon"]])
```

```python
# 🔍 Diagnóstico inicial do DataFrame integrado
import pandas as pd

print("📄 Dimensões:", df_full.shape)
print("\n🔢 Tipos de dados por coluna:")
print(df_full.dtypes)

print("\n🔍 Amostra das primeiras linhas:")
display(df_full.head(5))

print("\n🔍 Colunas com valores nulos:")
display(df_full.isna().sum()[df_full.isna().sum() > 0])

print("\n🧭 Amostra de colunas de coordenadas (para detectar formato incorreto):")
for c in [col for col in df_full.columns if "lat" in col.lower() or "lon" in col.lower()]:
    print(f"\n▶ {c}")
    print(df_full[c].dropna().astype(str).head(10).to_list())
```

```python
# 5. Salvar dataset unificado

import pandas as pd
import numpy as np

# Assuma que seu DataFrame está em df_full
df = df_full.copy()
```

```python
# ---------- 1) Funções auxiliares ----------
def _fix_coord(col: pd.Series, is_lat=True) -> pd.Series:
    """Converte coordenada em string (e.g. '-119283' ou '-619953.0') para float em graus.
       Detecta escala ×10000 e ajusta automaticamente."""
    s = col.astype(str).str.replace(",", ".", regex=False).str.strip()
    s = s.str.extract(r"(-?\d+(?:\.\d+)?)", expand=False)
    s = pd.to_numeric(s, errors="coerce")
    thr = 90 if is_lat else 180
    if s.dropna().abs().gt(thr).mean() > 0.5:
        s = s / 10000.0
    return s.astype("float64")
```

```python
# ---------- 2) Corrigir coordenadas ----------
if "LAT__lat" in df.columns:
    df["LAT__lat"] = _fix_coord(df["LAT__lat"], is_lat=True)
if "LAT__lon" in df.columns:
    df["LAT__lon"] = _fix_coord(df["LAT__lon"], is_lat=False)
```

```python
# ---------- 3) Garantir tipos dos indicadores ----------
INDICADORES = ["IND2","IND4","IND5","IND8","IND9","INF1","INF4-DL","INF4-UP"]
for c in INDICADORES:
    if c in df.columns:
        df[c] = pd.to_numeric(df[c], errors="coerce").astype("float64")
```

```python
# ---------- 4) Normalizar tipos do bloco LAT (onde existir) ----------
if "LAT__capital" in df.columns:
    df["LAT__capital"] = pd.to_numeric(df["LAT__capital"], errors="coerce").astype("Int64")
for c in ["LAT__uf","LAT__fuso_horario","LAT__chave_uf_nome","LAT__nome_mun"]:
    if c in df.columns:
        df[c] = df[c].astype("string")
for c in ["LAT__siafi_id","LAT__ddd"]:
    if c in df.columns:
        df[c] = pd.to_numeric(df[c], errors="coerce").astype("Int64")
```

```python
# ---------- 5) Criar chaves canônicas (se faltarem) ----------
if "uf" not in df.columns and "uf_soc" in df.columns:
    df["uf"] = df["uf_soc"]
if "nome_mun" not in df.columns and "nome_mun_soc" in df.columns:
    df["nome_mun"] = df["nome_mun_soc"]
```

```python
# ---------- 6) Remover colunas 100% nulas (ex.: IDHM__População total 20xx) ----------
null_all = [c for c in df.columns if df[c].isna().all()]
if null_all:
    print(f"🧹 Removendo colunas 100% nulas ({len(null_all)}): {null_all[:6]}{' ...' if len(null_all)>6 else ''}")
    df = df.drop(columns=null_all)
```

```python
# ---------- 7) Checagens rápidas ----------
print("📍 Coordenadas (após ajuste):")
for c in [x for x in ("LAT__lat","LAT__lon") if x in df.columns]:
    print(f" - {c}: dtype={df[c].dtype}, min={df[c].min()}, max={df[c].max()}")

print("\n🔎 Indicadores → tipos:")
print(df[INDICADORES].dtypes)
```

```python
# --- Diagnóstico rápido para confirmar o problema ---
df["URB__Rural"].value_counts(dropna=False).head(10)
```

```python
# Converter URB__Rural e outras colunas URB__* para numéricas
urb_cols = [c for c in df.columns if c.startswith("URB__")]
for c in urb_cols:
    df[c] = (
        df[c]
        .astype(str)
        .replace("-", np.nan)       # substitui hífen por NaN
        .replace("nan", np.nan)
        .replace("", np.nan)
    )
    df[c] = pd.to_numeric(df[c], errors="coerce")

print("✅ Colunas URB__ convertidas para float64:")
print(df[urb_cols].dtypes)

# Testar exportação novamente
OUT_PARQUET = "rqual_2022_integrado.parquet"
OUT_XLSX = "rqual_2022_integrado.xlsx"

df.to_parquet(OUT_PARQUET, index=False)
df.to_excel(OUT_XLSX, index=False)

print(f"\n💾 Exportado com sucesso:\n - {OUT_PARQUET}\n - {OUT_XLSX}")
print(f"Shape final: {df.shape[0]:,} linhas × {df.shape[1]:,} colunas")
```
